# Data Cleaning

Our data generally consists of two parts: graduation rate data and assessment achievement result data. Both these datasets come from the same source, and are therefore similarly broken down. The specific sets we are interested in are broken down by school and should theoretically contain every school in the country, labelled by state. The columns generally start with a top-level statistic (e.g. overall graduation rate), and proceed to go into detail (i.e. graduation rate broken down by demographic).

There is lengthy documentation that goes with each CSV file that explains their details. For example, there are many cells that contain NaN values or are blank. Cases where data for that particular combination of conditions was not available (e.g. a school may simply not have enough of a certain demographic to report useful data) are marked as NaN or n/a. Cases where there actually was no data reported at all are left blank. For our purposes, however, both these cases are virtually the same since we are interested in comparing percent rates—having inconsistencies would skew the data, so we simply choose to remove these rows from the datasets.

There are also cells marked ‘PS’. These are cases where data was available, but due to the extremely small number of reported datapoints in the category, it would be easy to personally identify students; these scores are overwritten by PS to protect student privacy. Again, for our purposes, it makes sense to just remove these rows.

Additionally, there are cells marked ‘GE’, ‘LE’, ‘GT’, and ‘LT’ followed by a number. These are also cases where student privacy is being protected due to low numbers, but instead of completely obfuscating the datapoint, an approximate range has been given instead. The four abbreviations stand for ‘Greater than or equal to’, ‘Less than or equal to’, ‘Greater than’, and ‘Less than’ respectively—a school’s scores for a particular demographic could be reported as being GT80, meaning that greater than 80% of students in that category achieved proficient scores. For these datapoints, we are choosing to make a conservative estimate (and we realize this will mean our analysis will be based on an approximation, but we think the error in the positive and negative directions should more or less average out) by simply picking the number from the term (i.e. GE80 becomes 80). This is close enough that we can conduct a useful analysis.

Further, we are extracting only the columns that are pertinent to our analysis. Since we are focused on homelessness and economic disadvantage, those are the columns we will be selecting from the datasets.

Lastly, there are cases where values are presented as ranges (i.e. 70-75)—we have chosen to simply take the mean for these values (another conservative estimate that should not effect the overall trends present in the data).

#### Utility function for cleaning

The code: 
- Converts all numerical strings into int 64's
- replaces all the comparison operator values: 'GE', 'LE', etc with NaN 
- finds the average of ranged values (i.e "40-46" = 48)
- replaces all other non numerical strings with NaN

```python
#Function to turn strings into usable numbers 
def clean(string): 
    try: 
        # Try to convert the passed in string to an int 64
        return np.int64(string) 
    except ValueError: 
        #Find the average of the ranged values 
        if '-' in string: 
            num1, num2 = string.split('-')
            num1 = np.int64(num1) 
            num2 = np.int64(num2) 
            string = np.int64((num1 + num2) / 2)
            return string

        # Replace the comparison operators
        if ('GE' in string) | ('LE' in string) | ('GT' in string) | ('LT' in string):
            return float('nan')

        #If there's anything else that's a string, just return NaN
        return float('nan')
```

## Graduation Rates

### 2018-2019 

This cell first imports the csv into a dataframe. It then pulls the columns that describes the state, total number of students that graduated (ALL_COHORT_1819), the total rate of graduation (ALL_RATE_1819), total number of homeless students that graduated (HOM_COHORT_1819), and the rate at which the homeless students graduated (HOM_RATE_1819) in that particular school. It then gets rid of rows with non-values that were inserted by the creators of the dataset for privacy reasons. Afterwards, it utilizes the converter function to convert the columns into usable data. It only runs it on the 'ALL_RATE_1819', 'HOM_COHORT_1819', and 'HOM_RATE_1819' rows because these are the only ones with the values described in the cleaner function's description. 

```python
#Read csv
gradRates_1819 = pd.read_csv('data/gradRates/gr2018-2019.csv')

#Grab the columns described above
gradRates_1819 = gradRates_1819[['STNAM','ALL_COHORT_1819','ALL_RATE_1819', 'HOM_COHORT_1819', 'HOM_RATE_1819']]

# Get rid of the values inserted for privacy (see data section for more details)
nonsense = ['', 'PS', '.']
gradRates_1819.replace(nonsense, float("NaN"), inplace=True)
gradRates_1819.dropna(inplace=True)

# Use the cleaner function explained at the start to clean columns that need it
colToClean = ['ALL_RATE_1819', 'HOM_COHORT_1819', 'HOM_RATE_1819']
for col in colToClean: 
    gradRates_1819[col] = gradRates_1819[col].apply(clean)

# Convert the numerical columns to int64's. 
gradRates_1819.dropna(inplace=True)
for col in colToClean: 
    gradRates_1819[col] = gradRates_1819[col].astype(np.int64)

#Check if any empty, 'PS', or NaN values exist in the df
print(('PS' in gradRates_1819) | ('' in gradRates_1819) | (gradRates_1819.isna().values.any()))
```

We can see that the data set now has our desired types and has shrunk considerably. Just based on quick glance of the All_rate and HOM_rate, we see that the HOM_rate is lower. 

Right now, the "all" student category includes homeless grad rates. For the sake of a better comparison, we're going to make two new columns: the number of non-homeless students that graduated (NOTHOM_cohort) and the rate of non-homeless students that graduated (NOTHOM_rate). This can be done by first finding the total number of non-homeless students that graduated which is simply the total number of students that graduated (ALL_COHORT) subtracted by the total number of homeless students that graduated (HOM_COHORT). Then we find the total number of students recorded in a given row by dividing the total number of students that graduated (ALL_COHORT) by the total graduation rate ('ALL_RATE'). Using the same logic, we find the total number of homeless students and then subtract the total number of students by the total number of homeless students to get the total number of non-homeless students. Then we find the rate of non homeless graduation by dividing the total number of non-homeless students that graduated from the total number of non-homeless students. 

```python
gradRates_1819['NOTHOM_cohort'] = gradRates_1819['ALL_COHORT_1819'] - gradRates_1819['HOM_COHORT_1819']

# Calculate the number of total non-homeless students 
gradRates_1819['NOTHOM_total'] = np.int64(np.round((gradRates_1819['ALL_COHORT_1819'] / (gradRates_1819['ALL_RATE_1819'] / 100)) - (gradRates_1819['HOM_COHORT_1819'] / (gradRates_1819['HOM_RATE_1819'] / 100)), 0))
gradRates_1819['NOTHOM_rate'] = np.int64(np.round((gradRates_1819['NOTHOM_cohort'] / gradRates_1819['NOTHOM_total']) * 100, 0))


#Cut out rows where rounding errors caused the not-homeless cohort rate equal a negative number 
gradRates_1819 = gradRates_1819[gradRates_1819['NOTHOM_rate'] > 0]

#Get rid of temp column 
gradRates_1819.pop('NOTHOM_total')

# Rearrange the columns 
col = gradRates_1819.pop('NOTHOM_cohort')
gradRates_1819.insert(3, col.name, col)
col = gradRates_1819.pop('NOTHOM_rate')
gradRates_1819.insert(4, col.name, col)

gradRates_1819.head()
```

#### Two Sample Kolmogorov Smirnov Test

```python
# Drop the total column because it's not a state
mean_gradRates = mean_gradRates.drop(index="Total")

#Perform the test 
hom_grad = mean_gradRates[['HOM_rate']].squeeze().sample(n=30, random_state=1)
nonHom_grad = mean_gradRates[['NOTHOM_rate']].squeeze().sample(n=30, random_state=1)
ks_2samp(hom_grad, nonHom_grad)
```

Our critical value $D_\alpha = c(\alpha)\sqrt{\frac{n_1 + n_2}{n_1 n_2}}$. 
We used sample sizes of 30 and with a $\alpha = .05$, $c(\alpha) = 1.36$.

When plugged into our formula, we get a $D_\alpha \approx .35115$. Our statistic was 0.833. Since 0.833 > .35115, we reject the null hypothesis and find that the samples are not drawn from the same distribution. Just as a note, our statistic was still greater than $D_\alpha$ when we calculated $D_\alpha$ using an alpha of 0.001 (0.503), so we can be very certain that these two samples do not come from the same distribution. 

#### One-tailed Two Sample T test

The populations were somewhat normally distributed and had enough observations to statisfy the Central Limit Theorem, so we thought it was okay to use a t-test to compare the homeless and non-homeless populations.

Start by ensuring that the variances of the two are the same by using the 4:1 ratio general practice:

```python
checkVar = False 
# Find which is larger and then ensure the 4:1 ratio is not exceeded
if (hom_grad.var() > nonHom_grad.var()): 
    checkVar = (hom_grad.var() / nonHom_grad.var())  < 4
else: 
    checkVar = (nonHom_grad.var() / hom_grad.var()) < 4
checkVar
```

 The above code is just an admittedly overengineered way of checking if the variances do not exceed the 4:1 ratio. Since the Boolean value is true, we can assume the variances do not exceed the 4:1 ratio and we can assume the variances are equal for our t test.

```python
t_val, p_val = ttest_ind(a=hom_grad, b=nonHom_grad)
t_val, p_val 
```

Since both samples had 30 observations, the total degrees of freedom was 60-2 = 58. Using the one-tailed critical value t-table at https://www.statisticshowto.com/tables/t-distribution-table/, we find that the critical t value is $\approx$ -1.671 at an alpha of .05. 

Our test statistic was -11.61 which is much less than -1.671. This indicates that the homeless graduation rate was significantly lower than that of the non-homeless student population. As a matter of fact, we could run the test with an alpha of .0005 (crit value $\approx$ -3.460) and still reject the null hypothesis.

# Limitations

The limitations with our project mostly have to do with the quality and quantity of data available, as well as more general limitations with our topic in general. 

Addressing the more general issue first, homelessness, by nature, is difficult to report accurately. This is even more true for unsheltered homeless people. [The Committee on Health Care for Homeless People](https://www.ncbi.nlm.nih.gov/books/NBK218229/) outlines common methods used to gain this data and discusses how difficult it is to just get an accurate estimate for a census, much less specific data like the graduation rates of homeless children. All that is to say that even if we are given all the graduation rates for homeless children in schools, these are only addressing those already enrolled in schools—there are potentially several children who are not enrolled in schools that are unaccounted for, making the numbers seem rather low.

This leads into the quantity of data available. Most states besides California have very little data available—even for homeless children enrolled in schools—to the point where it’s almost impossible to draw aggregate trends. Additionally, we originally wanted to include children in foster care in our analysis, but there was so little data available we were forced to scrap that section. 

Finally, out of the data that is available, a lot of its reliability is questionable. For example, there are certain rows in our CSV files where a manual calculation reveals that the total number of graduates in the row is somehow higher than the total number of students in each category, which is impossible. Further, because there is so little data available, a lot of it is obfuscated to protect the privacy of students involved, adding an additional layer of difficulty to the analysis.
