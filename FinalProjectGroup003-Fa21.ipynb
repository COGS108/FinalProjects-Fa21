# Web Scraping - NYT Bestsellers (Fiction/Non-Fiction)

```python
## WEB SCRAPING

# specify webpage we want to scrape 
### FICTION BEST SELLERS
wiki_fiction = 'https://en.wikipedia.org/wiki/Lists_of_The_New_York_Times_Fiction_Best_Sellers'
req_fiction = requests.get(wiki_fiction)
soup_fiction = BeautifulSoup(req_fiction.content, 'html.parser') # get contents of web page

### NONFICTION BEST SELLERS
wiki_nonfiction = 'https://en.wikipedia.org/wiki/Lists_of_The_New_York_Times_Nonfiction_Best_Sellers'
req_nonfiction= requests.get(wiki_nonfiction)
soup_nonfiction = BeautifulSoup(req_nonfiction.content, 'html.parser')
```

```python
## FICTION SET

# class list set -- this finds our wanted table
fiction_table = soup_fiction.find('table', {'class':"nowraplinks"})
#print(fiction_table)

#find all the links for every year
fiction_links = fiction_table.findAll('a')
print("\n")
#print(fiction_links)

#make a list of links by year
FictionYears = []
for link in fiction_links:
    FictionYears.append(link.get('href'))
    #remove junk titles and all years until 1942
    #in 1942 the national list was created, before that it was only NY area
del FictionYears[0:15]
#print(FictionYears)
    
#create a df
fiction_bestsellers = pd.DataFrame(columns=['Year', 'Title', 'Author', 'Genre', 'NumWeeks'])

# keep track of years
current_year = 1942

#go to every link and get the wikitable
for link in FictionYears:
    
    # webscrape each link
    year_link = 'https://en.wikipedia.org'+str(link)
    year_req = requests.get(year_link)
    year_soup = BeautifulSoup(year_req.content, 'html.parser')
    #print(year_link)
    
    # get the wikitable
    current_year_table = year_soup.find('table', {'class':"wikitable sortable"})
    
    # get the title of each 
    book_titles = year_soup.findAll('a')
    
    
    num_weeks = 1
    # Collecting Ddata
    for row in current_year_table.tbody.find_all('tr'):    
        # Find all data for each column
        columns = row.find_all('td')
    
        if(len(columns) == 3):
            title = columns[1].text.strip()
            author = columns[2].text.strip()
            fiction_bestsellers = fiction_bestsellers.append({'Year': current_year, 'Title': title, 'Author': author, 'Genre': 'Fiction', 'NumWeeks': num_weeks}, ignore_index=True)

        ## if the number of 'td' in the current 'tr' is 1, it means its the same book
        if (len(columns) == 1):
            num_weeks = num_weeks+1
            fiction_bestsellers.at['index', "NumWeeks"] = num_weeks
        else: num_weeks = 1
                     
    current_year = current_year+1
print("Scraping Complete")
## please note that the number of weeks as bestseller for the (last book only!) will not be available
fiction_bestsellers['NumWeeks'] = fiction_bestsellers.NumWeeks.shift(-1) 
fiction_bestsellers = fiction_bestsellers.dropna()
```

```python
## NONFICTION SET

# class list set -- this finds our wanted table
nonfiction_table = soup_nonfiction.find('table', {'class':"nowraplinks"})
#print(nonfiction_table)

#find all the links for every year
nonfiction_links = nonfiction_table.findAll('a')
#print(nonfiction_links)

#make a list of links by year
NonFictionYears = []
for link in nonfiction_links:
    NonFictionYears.append(link.get('href'))
    
    #remove junk titles and all years until 1942
    #in 1942 the national list was created, before that it was only NY area
    #####NOTE: THE NONFICTION SITE IS INCOMPLETE, DATA MISSING, WE WILL START AT 1991 FOR THIS SET
del NonFictionYears[0:64]

#rename the cols to match
nonfiction_bestsellers = pd.DataFrame(columns=['Year', 'Title', 'Author', 'Genre', 'NumWeeks'])

# keep track of years
current_year = 1991

#go to every link and get the wikitable
for link in NonFictionYears:
    
    # webscrape each link
    year_link = 'https://en.wikipedia.org'+str(link)
    year_req = requests.get(year_link)
    year_soup = BeautifulSoup(year_req.content, 'html.parser')
    # print(year_link)
    
    # get the wikitable
    current_year_table = year_soup.find('table', {'class':"wikitable"})
    
    num_weeks = 1
    # Collecting Ddata
    for row in current_year_table.tbody.find_all('tr'):    
        # Find all data for each column
        columns = row.find_all('td')
    
        if (len(columns) == 3):
            title = columns[1].text.strip()
            author = columns[2].text.strip()
            nonfiction_bestsellers = nonfiction_bestsellers.append({'Year': current_year, 'Title': title, 'Author': author, 'Genre': 'NonFiction', 'NumWeeks': num_weeks}, ignore_index=True)
        
        #in 2008 they started adding a 4th col
        if (len(columns) == 4):
            title = columns[1].text.strip()
            author = columns[2].text.strip()
            nonfiction_bestsellers = nonfiction_bestsellers.append({'Year': current_year, 'Title': title, 'Author': author, 'Genre': 'NonFiction', 'NumWeeks': num_weeks}, ignore_index=True)
    
        count = 0
        count = count+1
    
        ## if the number of 'td' in the current 'tr' is 1, it means its the same book
        if (len(columns) == 1):
            num_weeks = num_weeks+1
            nonfiction_bestsellers.at[count, "NumWeeks"] = num_weeks
        else: num_weeks = 1
            
    current_year = current_year+1
print("Scraping Complete")
## please note that the number of weeks as bestseller for the (last book only!) will not be available
nonfiction_bestsellers['NumWeeks'] = nonfiction_bestsellers.NumWeeks.shift(-1) 
nonfiction_bestsellers = nonfiction_bestsellers.dropna()
```

## Star Wars Trends vs Relevant Names
First we can visualize the google trends data for "Star Wars" from 2004. We can clearly see spikes in 2016, 2018 and 2020 when the movies in the most recent trilogy were released. Ideally if there's some correlation between popularity and names, we'd also see upticks in related names during these periods. Since baby name data is only in increments of full years, we need to first sum all rows of a given year into a single row to make the data fit together. We'll make a function to do this analysis and graphing for a given name to easy analyze and identify any number of names with this dataset.

```python
def nameVsStarWars(name, gender):
    sw_year_t = star_wars_trends.set_index(['month'], drop=True)
    
    # Make months into single row for year using max to choose the value
    sw_year_t = sw_year_t.resample('1YS').max()
    sw_year_t.index.names = ['year']
    name_data = baby_data[(baby_data['name'] == name) & (baby_data['year'] >= 2004) & (baby_data['gender'] == gender)]
    name_data = name_data[['count','year']].reset_index(drop=True)
    name_data.set_index(pd.to_datetime(name_data['year'], format='%Y'), drop=True, inplace=True)
    name_data_combined = sw_year_t[sw_year_t.index < pd.to_datetime('2021', format='%Y')]
    name_data_combined['count'] = name_data['count']
    name_data_combined['count'].fillna(0, inplace=True)
    name_data_combined = name_data_combined.reset_index()
    ax = sns.lineplot(data=name_data_combined, x='year', y='rel_pop',color='red')
    ax.set_ylabel("Relative Popuplarity")
    plt.legend(labels=['Relative Popularity'])
    ax2 = ax.twinx()
    ax2.set_ylabel(name + " Count")
    sns.lineplot(data=name_data_combined,x='year', y='count', ax=ax2,color='green')
    plt.legend(labels=[name + ' Count'],loc = 1)
    name_data_combined.head()
    return name_data_combined
```

Now lets get the data for a single name from the most recent trilogy. Let's pick Rey, as if we look at the baby name data, we can see that this is an extremely unpopular name for a female. The first instances of it were in 2016 (the year the first movie in the newest star wars trilogy was released).

```python
rey_data = nameVsStarWars('Rey','F')
```

It looks like the first instances of baby girls named Rey occured shortly after the release of 'The Force Awakens'. But we can notice that the data for the counts of baby names seems delayed about a year after the spikes in google trends popularity for the star wars franchise. Let exclude data prior to 2014 (as any star wars popularity in that timeframe would be prior to there being a character named Rey), and shift the baby name count back one year to more closely match the popularity.

```python
shifted_rey = rey_data
shifted_rey['count'] = shifted_rey['count'].shift(-1)
shifted_rey.dropna(inplace=True)
shifted_rey = shifted_rey[10:]

ax = sns.lineplot(data=shifted_rey, x='year', y='rel_pop',color='red')
ax.set_ylabel("Relative Popuplarity")
ax2 = ax.twinx()
ax2.set_ylabel('Rey' + " Count")
sns.lineplot(data=shifted_rey,x='year', y='count', ax=ax2,color='green');
```

```python
#Calculating normalized slopes of the count/popularity data
slopes = shifted_rey.iloc[0:].diff()
slopes.dropna(inplace=True)
maxVal = slopes['count'].max()
slopes['count'] = slopes['count'].div(maxVal / 100)
slopes
```

```python
# carry out linear regression
outcome, predictors = patsy.dmatrices('count ~ rel_pop', slopes)
mod_log = sm.OLS(outcome, predictors)
fit_res = mod_log.fit()
print(fit_res.summary())
```

In doing linear regression, we can see that there's a positive correlation between the slope of the relative popularity of the star wars search terms and the normalized (between 0 and 100 like the pop) increase in the number of female 'Rey's born in each year from 2014 onward. However, given how few data points there are, a strict statistical correlation is not all that meaningful. The most relevant piece of information we can gather from this data is that there is clearly a correlation between the release and popularity of each of the new star wars movies and the name Rey. It's also extremely compelling that prior to the year following the movie's release, there had been between 0-5 female Rey's born in America per year dating all the way back to 1880. # Percentage Baby Name Change After Song Release

In order to capture the impact of the song release as much as possible, the percentage change is calculated as follows:

$$ ((BabyNameCount_{Release Year} + BabyNameCount_{Release Year+1}) / (2*BabyNameCount_{ReleaseYear-1}) - 1) * 100 $$

The BabyName Count from the release year and the year thereafter are averaged and compared to the BabyName Count from the year prior to release. By limiting the range of time to  1 year from the song release, we can more strongly attribute the effect on BabyName count to the release of the song instead of an unknown external factor. For this analysis, only names with a matching song title are used to maximize the likelihood of causation.

```python
def getMusic(title):
#     Helper function to get music information based on song title
    return music_top100[music_top100.index == title]
```

```python
def NameMusicIntersection():
#     Calculates the intersection between baby names and song titles
    names = baby_data['name'].tolist()
    music_test = set([x.lower() for x in music_top100.index.tolist()])
    names = set([x.lower() for x in names])
    names = names.intersection(music_test)
    return list(names)
```

```python
def CalcPNC(names):
#     Function to calculate percentage baby name change
    PNC = []
    Genres = []
    PeakPositions = []
    NumWeeksOnChart = []
    for name in names:
        [percentChange, Genre, PeakPosition, woC] = PercentNameChange(name.lower(),name.lower())
#         Append values to list
        PNC.append(percentChange)
        Genres.append(Genre)
        PeakPositions.append(PeakPosition)
        NumWeeksOnChart.append(woC)
    return [PNC, names, Genres, PeakPositions, NumWeeksOnChart]
```

```python
def NameSkip(name):
#     Name to skip
    skip = ['mercy','real']
    return (name.lower() in skip)
```

```python
def PercentNameChange(name,title):
    #Names like Mercy have 5 different entries with overlapping years
    UName = name[0].upper()+name[1:]
    if NameSkip(name):
        return [np.nan, np.nan, np.nan, np.nan]
    #     Set song titles to lowercase
    music = music_top100.reset_index()
    music['Title'] = music.reset_index()['Title'].str.lower()
    music = music.set_index('Title')
#     Get aggregated baby name counts
    df = NoGenderBabyData[NoGenderBabyData['name'] == UName]
    WeeksonChart = 1
    if(name.lower() == 'nann'):
        releaseYear = 1998
        Genre = 'Rap'
        PeakPosition = 62
    elif(name.lower() == 'faith'):
#         George Michael's Faith from 1987 randomly appears for one week in 2017 (RIP George Michael)
        releaseYear = 1987
        Genre = 'Blue-Eyed Soul'
        PeakPosition = 1
    elif(name.lower() == 'fortunate'):
#         Song was released before the dataset was created
        releaseYear = 1999
        Genre = 'R&;B'
        PeakPosition = 4
    else:
#         Get releaseYear, Genre, and PeakPosition
        releaseYear = music[list(music.index == title) & np.isnan(music['# of Weeks on Chart'])]['Week'].dt.year.values[0]
        temp = music[list(music.index == title) & np.isnan(music['# of Weeks on Chart'])]
        Genre = temp.iloc[0].Genre
        Artist = temp.iloc[0].Artists
        PeakPosition = music[np.logical_and(list(music.index == title),list(music['Artists'] == Artist))].dropna()['Peak Position'].min()
        WeeksonChart = music[np.logical_and(list(music.index == title),list(music['Artists'] == Artist))].dropna()['# of Weeks on Chart'].max()
        if(np.isnan(PeakPosition)):
            PeakPosition = music[np.logical_and(list(music.index == title),list(music['Artists'] == Artist))]['Weekly Rank'].values[0]
        if(np.isnan(WeeksonChart)):
            WeeksonChart = music[np.logical_and(list(music.index == title),list(music['Artists'] == Artist))]['# of Weeks on Chart'].values[0]
    counts = [-1,-1,-1]
    years = set([releaseYear,releaseYear+1,releaseYear-1])
#     Check if there is baby name data for the years needed for calculation
    if set([releaseYear,releaseYear+1,releaseYear-1]).issubset(set(df['year'])):    
        counts[0] = df.set_index('year').loc[releaseYear].values[1]
        counts[1] = df.set_index('year').loc[releaseYear+1].values[1]
        counts[2] = df.set_index('year').loc[releaseYear-1].values[1]
        percentChange = 100 * (((counts[0] + counts[1]) / (2 * counts[2])) - 1)
        return [percentChange, Genre, PeakPosition, WeeksonChart]
    else:
        return [np.nan, np.nan, np.nan, np.nan]  
```

```python
x = CalcPNC(NameMusicIntersection())
d = {'PNC':x[0],'Title':x[1]}
df = pd.DataFrame(data=d).dropna()
df.sort_values(by='PNC')
df['PNC'].mean()
```

The average percent name change correlated with song release was +19.57% overall. This gives good evidence that the song releases had an impact on baby names because the percentage changes should average out to about 0 if there was no effect.

# Peak Position Correlated with Percentage Name Change

In this next section we'll see if the relative popularity of the songs affects the magnitude of percent change. Peak position on the Music Top 100 chart is used to gauge relative popularity because we can assume that more popular songs climbed to a higher spot on the charts.

```python
d = {'PNC':x[0],'Title':x[1],'Genre':x[2], 'PeakPosition':x[3]}
df = pd.DataFrame(data=d).dropna()
df['PNC'] = abs(df['PNC'])
df.plot(x='PeakPosition',y='PNC',kind='scatter',ylabel='Percentage Name Change',xlabel='Peak Position on the Music Top 100 Chart');
```

```python
# carry out linear regression
outcome, predictors = patsy.dmatrices('PNC ~ PeakPosition', df)
mod = sm.OLS(outcome, predictors)
fit_res = mod.fit()
print(fit_res.summary())
```

According to our data we see that there is little to no correlation between peak position on the Music Top 100 chart and percentage name change (r = 0.002). There are multiple explanations for this. Firstly, songs that make the Music Top 100 chart are already in the top 0.01 percentile of songs so their influence on the general public is going to be about the same. Secondly, our dataset is limited to names that have matching song titles such as 'Havana' or 'Warrior' which have low baby name counts to begin with (65 and 12 respectively for 2020) compared with names like 'Roxanne' which has a name count in the hundreds. This leads to noise being added to the data because names with low counts have a larger PCN variance than names with higher counts. Let's see what happens when we look at the number of weeks on the top 100 chart to represent relative popularity.

This is an improved analysis over using Peak Position to represent relative popularity (r=0.055), but it still suffers from the same caveats as the previous analysis. Namely, these are already the most popular songs because they're on the billboard top 100, and we're limited to songs with matching names. Regardless, you can still see a slight positive trend as the number of weeks on the chart increases. Overall, number of weeks on chart is a better indicator of impact on baby name count increase than peak position; however, it's overshadowed by the base popularity a song requires to make it on the chart to begin with.
