# Research Question

What effect did population density have on the nature of rush hour collisions, defined as the intense period of traffic that occurs in between 6-9 A.M. and 4-7 P.M., in California from 2015-2020?

<a id='background'></a>

## Background & Prior Work

Rush hour is one of the defining traits of automotive transport. Those needing to commute during these hours, usually defined as from 6 to 9 in the morning and 4 to 7 at night, are resigned to spending long hours inching forward in heavy congestion, while others do their best to avoid driving at these times. Unsurprisingly, given the high density of cars, rush hour times have the highest number of collisions of any part of the day (1). 

However, while the overall distribution of crashes in a day should surprise no one, we became interested in looking deeper at how these overall trends are reflected across areas of different population density. We started our investigation with the theory that the nature of rush hour collisions would vary significantly between the densest areas and the least. Background research revealed that nationally there are far more collision deaths in rural areas than urban ones and that there are significant differences in the nature and cause of urban and rural traffic deaths (2). 

Traffic fatalities are certainly an important metric, but are a limiting perspective to take. We theorized that despite the high number of collisions in urban rush hour, there would be fewer fatalities per collision than in rural areas where there is less congestion and the speeds are higher. By expanding  our analysis to collisions, not exclusively deaths, we feel we are better able to describe the urban-rural divide during rush hour and characterize the effect of population density on rush hour collisions. To determine this, the target of our analysis is the last five years of California traffic collisions divided by the 58 counties of the state. 

References:
- 1) https://injuryfacts.nsc.org/motor-vehicle/overview/crashes-by-time-of-day-and-day-of-week/
- 2) https://www.iihs.org/topics/fatality-statistics/detail/urban-rural-comparison

# Hypothesis


Our hypothesis attempts to relate four conditions of collisions: distracted driving, hit and runs, fatality rate, and weather and fatalities, with population density. 

Due to the slow speed of urban rush hour traffic, we expect that more drivers will feel safe to look at their phones, and as such we will observe a higher rate of distracted driving collisions in denser areas. Related to the density of urban traffic, we theorize that there should be fewer hit and runs in denser areas as there are more witnesses and less space to flee. 

We expect areas of lower population density to experience less rush hour traffic and, as such, higher driving speeds. Because of this, we hypothesize that this will result in a higher collision fatality rate in less dense areas. The higher speeds achievable in less dense areas should also result in an even higher fatality rate in bad weather than denser areas.

# Data Cleaning

Once our initial 10 gigabyte dataset was converted into usable CSV files we began cleaning our data. We start by reading all our datasets into dataframes and taking a look at what will be our main dataset, 'df_collisions,' which will henceforth be refered to as collisions.

````python
# Read in counties data
df_counties = pd.read_csv("csvData.csv")

# Read in collisions data
df_collisions = pd.read_csv('collisions1.csv')
df_collisions = pd.concat((df_collisions, pd.read_csv('collisions2.csv') ))
df_collisions = pd.concat((df_collisions, pd.read_csv('collisions3.csv') ))

# Read in Parties and Victims data
df_parties = pd.concat((pd.read_csv('parties1.csv'), pd.read_csv('parties2.csv')))
df_victims = pd.read_csv('victims.csv')

df_collisions.head()
````

We see an unexpected column we dont want named 'Unnamed:0'. This column is also present in the parties and victims dataframes so we will go ahead and remove it from all three. We reset the index on collisions and parties to account for the fact that they were multiple CSV files concatenated together. We also drop 3 columns from our dataset. These columns were initially grabbed with our SQL queries on our large dataset but ultimately we decided they were not of interest to our project. Instead of recreating our CSV's we decided it would be simpler to drop them from the dataset. After these steps we take a look at collisions again.

````python
# Remove unexpected column 'Unnamed: 0'
df_collisions.drop('Unnamed: 0', axis=1, inplace=True)
df_parties.drop('Unnamed: 0', axis=1, inplace=True)
df_victims.drop('Unnamed: 0', axis=1, inplace=True)

# Reset index to account for multiple CSV files in single dataframe
df_collisions.reset_index(drop=True, inplace=True)
df_parties.reset_index(drop=True, inplace=True)

# Drop unused columns
df_collisions = df_collisions.drop(columns = ['primary_collision_factor','alcohol_involved']) #'collision_severity'

df_collisions.head()
````

Let's check the size of our dataset to see how many collisions we are working with.

````python
print(len(df_collisions))
````

We elect to get rid of any point of data that contains a null value in any of the columns we will be inspecting. We decided this was acceptable to do since we have such a large dataset with 2.7 million data points and this process only removes 35,000. We see the number of null values in each of the columns below:

````python
df_collisions.isnull().sum()
````

We updated collisions to get rid of all these null values. As we do this, we're also saving all the Case IDs of these data points to remove them from the parties and victims datasets as well since the data will be incomplete if we don't do so. Finally, we'll check the amount of Case IDs we're deleting to ensure it's not more than expected.

````python
# Remove null values and store their Case IDs - collision_time
del_caseIds = df_collisions[ df_collisions['collision_time'].isnull() ]['case_id']
df_collisions = df_collisions[df_collisions['collision_time'].notna()]

# '' - weather_1
del_caseIds = pd.concat((del_caseIds, df_collisions[ df_collisions['weather_1'].isnull() ]['case_id'])) 
df_collisions = df_collisions[df_collisions['weather_1'].notna()]

# '' - 'hit_and_run
del_caseIds = pd.concat((del_caseIds, df_collisions[ df_collisions['hit_and_run'].isnull() ]['case_id'])) 
df_collisions = df_collisions[df_collisions['hit_and_run'].notna()]

print(len(del_caseIds))
````

Next, we remove those selected Case IDs from the parties and victims dataset. From here, we must deal with the issue of the parties dataset's ability to have multiple entries with the same Case ID, so we choose to merge all the cases together into one Case ID and add up the cellphone in use column. By doing this, we make it so if any party in the collision had cellphone in usage as true, we ensure that the entire Case ID has a cellphone in use. We do this by changing any value >1 in the cellphone column to 1, representing that unique Case ID had a cellphone involved in that accident. We then merge this dataset into collisions to incorporate the cellphone column. Finally we check how many null values this created and then take a look at our merged dataset.

````python
# Remove Case IDs from Parties and Victims that were removed from Collisions
df_parties = df_parties[ ~df_parties['case_id'].isin(del_caseIds) ]
df_victims = df_victims[ ~df_victims['case_id'].isin(del_caseIds) ]
df_collisions.sort_values(['collision_date', 'collision_time'], inplace=True)

# Combine all parties with matching Case IDs by adding cellphone_in_use values
df_parties = df_parties.groupby('case_id').agg({'cellphone_in_use':'sum'})
df_parties[df_parties['cellphone_in_use'] != 0] = 1.0

# Merge Collisions and Parties to add cellphone_in_use values
df_merged = pd.merge(df_collisions, df_parties, how='left', on='case_id')

print("null values in cellphone_in_use:", df_merged['cellphone_in_use'].isna().sum())
df_merged.head()
````

Next we look at our counties dataset. From this dataset all we want is the population density of each county in California so we drop other columns that do not contain this information. We change the column name and the format of the county names within the column to match that of the collisions dataset. We look at this dataset to see that the county format matches and what the population density variables look like.

````python
# Remove unwanted columns
df_counties.drop(["pop2021","GrowthRate"],axis=1, inplace=True)

# Match county name format with Collisions
df_counties['CTYNAME'] = df_counties['CTYNAME'].str.lower()
df_counties['CTYNAME'] = df_counties['CTYNAME'].str.replace(' county', '')
df_counties = df_counties.rename(columns= {'CTYNAME': 'county_location'})

df_counties.head()
````

We merge counties with our collisions dataset in order to give each Case ID a population density based on where the case was filed. At this point this dataset contains all the information we are going to look at. All we need to do is make a few final quality of life changes.

````python
df_merged = pd.merge(df_merged, df_counties, how = 'left', on='county_location')
df_merged.head()
````

The housekeeping steps we take to clean the dataset are:
 - Change column 'weather_1' to 'weather'.
 - Change 0 and 1's in cellphone column to False and True respectively.
 - Combine the Date and Time columns into one functional DateTime object and store it in a new column.
 - Drop Date and Time columns since they are now redundant.
 - Set 'df' equal to our final dataset to begin analysis.

````python
# Changing weather_1 column name to weather
df_merged.rename(columns=({'weather_1':'weather'}), inplace=True)

# Changing 0 and 1 values to False and True in alcohol_involved and cellphone_in_use
df_merged['cellphone_in_use'] = df_merged['cellphone_in_use'].replace([0,1],[False,True])
df_merged.head()

# Combine date and time into a single column and make it a datetime object
df_merged['collision_dateTime'] = df_merged['collision_date'] + " " + df_merged['collision_time']
df_merged['collision_dateTime'] = pd.to_datetime(df_merged['collision_dateTime'])

# Drop date and time
df_merged = df_merged.drop(columns = ['collision_date', 'collision_time'])

df = df_merged
````

Lastly we add 'rush-hour-time-bool' to every data point representing if the collision occurred during Rush Hour. Anytime from 6-9 A.M. or 4-7 P.M. was considered rush hour for the purpose of this project. We take a look at the dataset now to see if anything else needs to be cleaned.

````python
# Create rush hour bool and set equal based on collision_dateTime
df['rush-hour-time-bool'] = False
df.loc[(((df.collision_dateTime.dt.hour >= 6) & (df.collision_dateTime.dt.hour < 9)) | 
        ((df.collision_dateTime.dt.hour >= 16) & (df.collision_dateTime.dt.hour < 19))),'rush-hour-time-bool'] = True
df.head()
````

The data appears clean so now we are ready to analyze!

# Data Analysis & Results

### Collisions Based on Rush Hour and Density

First, we are trying to find about how population density affect collisions on rush hour. We start plotting a graph that with relationship between hours and total collision

````python
fig,ax = plt.subplots(figsize = (12,4))

df.groupby(df['collision_dateTime'].dt.hour)['case_id'].count().plot(kind = 'line')

plt.margins(x=0)
plt.annotate('Morning Rush Hour', xy = (6,175000), xytext = (1,150000), arrowprops = dict(arrowstyle = "->", color="black"))
plt.annotate('Evening Rush Hour', xy = (16,75000), xytext = (12,100000), arrowprops = dict(arrowstyle = "->", color="black"))
plt.xticks(ticks = [0,3,6,9,12,15,18,21], labels = ['12 A.M.','3 A.M.', '6 A.M.', '9 A.M.', '12 P.M.','3 P.M.', '6 P.M.','9 P.M.'])
plt.xlabel("Hour of Day")
plt.ylabel('Number of Collisions')
plt.title("Total Automobile Collisions by Hour Throughout California from 2015-2020")
plt.style.use('ggplot')

rushHourTimes = [6,9,16,19]
for x in rushHourTimes:
    plt.axvline(x=x, color = 'gray', linestyle = '--');
````

This graph showcases the relationship between rush hour and total collisions throughout California. The total number of collisions is lowest late at night, which is expected given there are less vehicles on the road. The total number of collisions steadily rises and reaches its peak right after evening rush hour, at which point it begins the decline as night falls.

````python
# Hourly Collisions in Rush vs non-Rush Hour
df_county = df

densityCollisions = []
for popDensity, group in df_county.groupby(df_county['popDensity']):
    rush_hour = group[group['rush-hour-time-bool']==True]['case_id'].count()/6
    normal_hour = group[group['rush-hour-time-bool']==False]['case_id'].count()/18
    densityCollisions.append((popDensity,rush_hour,normal_hour))
        
df_dc = pd.DataFrame(densityCollisions)
df_dc.columns = ['popDensity', 'rushCollisions','normalCollisions' ]

fig,ax = plt.subplots(figsize = (12,8),ncols = 1, nrows = 2)

plt.subplot(2,1,1)
plt.scatter( x='popDensity', y='rushCollisions', data=df_dc,color='red', label = 'Rush Hour Collisions')
plt.scatter( x='popDensity', y='normalCollisions', data=df_dc, color = 'blue', label = 'All Other Collisions')
plt.xlabel('County Population Density')
plt.ylabel("Number of Collisions")
plt.title('Hourly Collisions in Rush vs non-Rush Hour with Los Angeles')
plt.legend(loc='best')

# Remove Los Angeles
df_dc = df_dc[:len(df_dc) - 1]

plt.subplot(2,1,2)
fig.subplots_adjust(hspace=0.4)
plt.scatter(x='popDensity', y='rushCollisions', data=df_dc,color='red', label = 'Rush Hour Collisions')
plt.scatter(x='popDensity', y='normalCollisions', data=df_dc, color = 'blue', label = 'All Other Collisions')
plt.xlabel('County Population Density')
plt.ylabel("Number of Collisions")
plt.title('Hourly Collisions in Rush vs non-Rush Hour Without Los Angeles')
plt.legend(loc='best');
````

This graph demonstrates the positive relationship between population density and the difference between rush hour and non-rush hour collisions. As counties become more dense, they have a relatively higher number of rush hour collisions than less dense counties.

````python
# Distracted driving (cell-phone-usage) and not distracted driving, normalized by hour, During Rush hour vs Not Rush hour
df_phone_usage = df

df_phone = df
# get rush hour collisions
df_phone = df_phone[df_phone['rush-hour-time-bool'] == True]
# get total number of collisions during rush hour per county
plot_this = []
for popDensity, group in df_phone.groupby(df_phone['popDensity']):
    # rush hour data per county
    rush_hour = group[group['rush-hour-time-bool']==True]
    
    # total collisions during rush hour per county
    rush_count = rush_hour['case_id'].count()
    
    # total collisions during rush hour with phone usage per county
    phone_count = rush_hour[rush_hour['cellphone_in_use'] == True]['case_id'].count()
    phone_effect = phone_count/rush_count
    plot_this.append((popDensity, phone_effect))

plot_this = pd.DataFrame(plot_this)
plot_this.columns = ['popDensity', 'phone_effect']

fig,ax = plt.subplots(figsize = (12,8))

plt.subplot(2,1,1)
plt.scatter(x='popDensity', y='phone_effect', data=plot_this, c='popDensity', cmap='inferno', label='Rush Hour Collisions')
plt.colorbar()
plt.xlabel("Population Density of County")
plt.ylabel('% of Collisions')
plt.title('Phone Induced Rush Hour Collisions By County')

# Remove Los Angeles
plot_this = plot_this[:len(plot_this) - 1]

plt.subplot(2,1,2)
fig.subplots_adjust(hspace=0.4)
plt.scatter(x='popDensity', y='phone_effect', data=plot_this, c='popDensity', cmap='inferno', label='Rush Hour Collisions')
plt.colorbar()
plt.xlabel("Population Density of County")
plt.ylabel('% of Collisions')
plt.title('Phone Induced Rush Hour Collisions By County Without Los Angeles');
````

````python
# Ratio of accidents during rush hour that are hit and run compared to non rush hour - Stacked Bar Chart
total_collisions = df_hit_and_run.groupby(df_hit_and_run['collision_dateTime'].dt.hour)['case_id'].count()
total_norun = df_nohit.groupby(df_nohit['collision_dateTime'].dt.hour)['case_id'].count()
total_mis = df_misdemeanor.groupby(df_misdemeanor['collision_dateTime'].dt.hour)['case_id'].count()
total_fel = df_felony.groupby(df_felony['collision_dateTime'].dt.hour)['case_id'].count()

columns = ['collision_dateTime', 'total_collisions', 'no_hit_and_run', 'misdemeanor', 'felony']
df_run_proportion = pd.DataFrame(columns = columns)

df_run_proportion['total_collisions'] = total_collisions
df_run_proportion['no_hit_and_run'] = total_norun
df_run_proportion['misdemeanor'] = total_mis
df_run_proportion['felony'] = total_fel

df_run_proportion['no_hit_and_run'] = df_run_proportion['no_hit_and_run'] / df_run_proportion['total_collisions']
df_run_proportion['misdemeanor'] = df_run_proportion['misdemeanor'] / df_run_proportion['total_collisions']
df_run_proportion['felony'] = df_run_proportion['felony'] / df_run_proportion['total_collisions']

df_run_proportion = df_run_proportion.drop(columns = [ 'no_hit_and_run', 'total_collisions', 'collision_dateTime'])
df_run_proportion.plot( kind = 'line', color={"misdemeanor": "blue", "felony": "red"})

rushHourTimes = [6,9,16,19]
for x in rushHourTimes:
    plt.axvline(x=x, color = 'gray', linestyle = '--')

plt.margins(x=0)
plt.annotate('Morning Rush Hour', xy = (6,0.1), xytext = (2,0.1), arrowprops = dict(arrowstyle = "->", color="black"))
plt.annotate('Evening Rush Hour', xy = (16,0.22), xytext = (12,0.22), arrowprops = dict(arrowstyle = "->", color="black"))
plt.rcParams['figure.figsize'] = (12,4)
plt.xticks(ticks = [3,6,9,12,15,18,21], labels = ['3 A.M.', '6 A.M.', '9 A.M.', '12 P.M.','3 P.M.', '6 P.M.','9 P.M.'])
plt.xlabel("Hour of Day")
plt.ylabel('Frequency')
plt.title("Frequency of Hit and Runs as a Portion of All Collisions Throughout California");
````

Despite total misdemeanor and felony hit and runs going up during the day, the frequency actually goes down. This has to do with the fact that there are just more cars on the road and a larger sample size.

In California, hit and runs that cause property damage are classified as misdemeanors while hit and runs that cause harm or a death are considered felonies. We can see that during the time of day when most cars are on the road, the frequency of misdemeanor and felony hit and run collisions remain somewhat constant. From this, we think that rush hour doesn't necessarily have an impact on frequency of misdemeanor or felony hit and run collisions, and rather it is due to more total cars being on the road during the daytime.

````python
# Hit and Run collisions during rush hour as a percentage of total rush collisions
df_fatal = df
# get rush hour collisions
df_fatal = df_fatal[df_fatal['rush-hour-time-bool'] == True]
# get total number of collisions during rush hour per county
plot_this = []
for popDensity, group in df_fatal.groupby(df_fatal['popDensity']):
    # rush hour data per county
    rush_hour = group[group['rush-hour-time-bool']==True] 
    # total collisions during rush hour per county
    rush_count = rush_hour['case_id'].count()
    # hit and run rush hour collisions count
    hR_count = rush_hour[rush_hour['hit_and_run'] != 'not hit and run']['case_id'].count()
    hR_rate = hR_count/rush_count
    plot_this.append((popDensity, hR_rate))
plot_this = pd.DataFrame(plot_this)
plot_this.columns = ['popDensity', 'hit_run_rate']

fig,ax = plt.subplots(figsize = (12,8))

plt.subplot(2,1,1)
plt.scatter(x='popDensity', y='hit_run_rate', data=plot_this, c='popDensity', cmap='viridis', label='Collision Hit and Run Rate')
plt.colorbar()
plt.xlabel("Population Density of County")
plt.ylabel('Frequency of Hit and Run')
plt.title('Rush Hour Hit and Run Collision Rate by County');

plot_this = plot_this[:len(plot_this) - 1]

fig.subplots_adjust(hspace=0.4)
plt.subplot(2,1,2)
plt.scatter(x='popDensity', y='hit_run_rate', data=plot_this, c='popDensity', cmap='viridis', label='Collision Hit and Run Rate')
plt.colorbar()
plt.xlabel("Population Density of County")
plt.ylabel('Frequency of Hit and Run')
plt.title('Rush Hour Hit and Run Collision Rate by County Without Los Angeles');
````

Our initial hypothesis was that hit and run incidents would be less common during rush hour in denser areas since more cars on the road means more witnesses, close-quarters driving, and less options for running away from the incident. 

Based on comparing the rate of hit and runs during rush hour by county population density, we see that there is actually a positive correlation between the rate of hit and runs and popultion density. Interestingly, we observe a fairly steady rate of hit and runs above the low population density of around 300 people per square mile. These results indicate that, aside from in the least populated areas, there are proportionally near the same amount of hit and runs everywhere, a finding contrary to our hypothesis.

# Ethics & Privacy

Our primary dataset is from the government created, and publically available, California Traffic Collision Statewide Integrated Traffic Records System. The second dataset is from world population review and the data is from US Census Bureau. Both of the datasets are publicly sourced data sets with no personally identifiable information, so we don’t anticipate any privacy concerns with obtaining or utilizing the datasets necessary for our project.   

The dataset from the US Census Bureau can be biased since some people didn’t fill out the survey, so that we may not have an accurate population. However, we are using the density instead of population, which we don’t think that this potential bias invalidates the data we are working with. We don’t have a method to reduce the bias in our data since the US Census Bureau already is the most reliable source for population in America.   

The dataset from SWITRS could potentially be biased since it is the product of human reporting. The data is dependent on the accuracy of the report given by police officers. Especially when we are using data points such as cellphone use which officers may not have firsthand knowledge of, there is the potential for favored groups to be absolved of responsibility. This potential bias could potentially prove to have an effect on the validity of our research; however, there is not an alternative source for this information and it's difficult to imagine reporting bias at the scale needed to invalidate our results. That being said, we would ideally be able to correct for this potential source of error, we just simply don't see a way to go about that. 

Given the anonymity afforded by the dataset and its collector, we don't forsee any privacy issues stemming from our project.
