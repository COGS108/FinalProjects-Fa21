## Background & Prior Work

We are trying to investigate what impacts the crime rate in different countries, specifically the rate of female murder victims versus male murder victims. Upon initial research, we came across the Global Gender Gap report, which we used as the major guide of our project. The Global Gender Gap Index benchmarks the evolution of gender-based gaps among four key dimensions (Economic Participation and Opportunity, Educational Attainment, Health and Survival, and Political Empowerment) and tracks progress towards closing these gaps over time” (Global Gender Gap Report). A higher Global Gender Gap Index (closer to 1)indicates that there are less gap, that male and females are more equal(Global Gender Gap Report, 2021). 

Based on our common sense and previous knowledge, we hypothesized that gender equality correlates with female and male murder victim counts. In places where females and males are more equal, the number of female victims will decrease. However, upon further research, we find out it is not so clear. For example, the factor of class may also play an important role.  In a study, researchers looked at 238 U.S. cities and their rape rates in Unifirm Crime Reports, comparing them to the relative and absolute women’s status in economic, educational, employment, and occupational spheres. Surprisingly, they found out that “greater gender equality was related to higher rape rates, but higher absolute status of women (which often accompanies higher gender equality) was associated with lower rape rates”(Martin & Vieraitis, 2006). Moreover, in a study done in 2021 by Chu, Hebenton and Toh, they looked at the relationship female offending and societal gender equality using the Global Gender Gap Index, and their findings shows that “ countries with a narrower gender gap in the Global Gender Gap indexes of political empowerment were likely to have a higher female prosecution rate, and that political empowerment was also significantly associated with female conviction rates as well as rates of property offending” (Chu, Hebenton & Toh, 2021).As a result, we found our asked question is more complicated than we thought. Many factors impact the rate of female and male victim rates. We think in order to get a holistic perspective on our topic, we will look at more factors and see which of them contribute the most to the dependent variable.

Besides looking at the Gender equality index, we also included the Gini index and HDI(Human Development index). The Gini index is a measure of statistical dispersion intended to represent the income inequality or the wealth inequality within a nation or a social group. The Gini coefficient measures the inequality among values of a frequency distribution (for example, levels of income). A Gini coefficient of zero expresses perfect equality, where all values are the same (e.g. where everyone has the same income). A Gini coefficient of one (or 100%) expresses maximal inequality among values(United Nations). We want to use the Gini index as another indicator of inequality that can possibly explain female vs male victims. 

According to the definition given by the United Nations, “The Human Development Index (HDI) is a summary measure of average achievement in key dimensions of human development: a long and healthy life, being knowledgeable and having a decent standard of living. The HDI is the geometric mean of normalized indices for each of the three dimensions”. We want to use the HDI as a general indicator of the condition and quality of life in countries, since crime rates in general can be caused by societal conditions.

# Data Cleaning

## Gender Equality Index Dataset  

From https://en.wikipedia.org/wiki/Global_Gender_Gap_Report and the World Economic Forum (WEF)

The raw dataset is relatively clean aside from the columns. To get the dataset usable, the current column headers should be removed since they contain metadata and the first row should be the headers. Then the column names should be formatted to be easily referenced. There are unused columns that will be removed like the Region and older years from 2006-2009. The observations are normalized based off of an index ranging from 0-1 so the observations are clean. We sorted the dataset based off of the most recent year to observe the degree of gender equality in each country.

```python
# Import raw dataset from the csv
df_gei = pd.read_csv("./gender-equality-index.csv")

# Drop the first row containing metadata
metadata = df_gei.iloc[0]
df_gei = df_gei.rename(columns=metadata).drop([0])

# Remove region and old data from 2006-2010 
df_gei = df_gei.drop(df_gei.columns[list(range(1,7))], axis=1)

# Remove recent data from 2020-2021
df_gei = df_gei.drop(df_gei.columns[[7,8]], axis=1)

df_tmp = pd.DataFrame(columns=['Country', 'GEI'])
df_tmp['Country'] = df_gei['Country']
for i, r in df_gei.iterrows():
    for col, val in enumerate(reversed(r)):
        if(val==val and col != len(r)-1):
            df_tmp.loc[i,'GEI'] = val
            break
            
df_gei = df_tmp.dropna().reset_index(drop=True)
df_gei['GEI'] = pd.to_numeric(df_gei['GEI'])
df_gei
```

## International Homicide by Female Dataset

From https://data.worldbank.org/indicator/VC.IHR.PSRC.FE.P5

The data set contains the number of female homicide victims per 100,000 females. The dataset is relatively clean aside from the NaN observations and columns that contain metadata. The first few columns and rows contain metadata that should be removed. The third row has the header we want so we will replace the headers with that row. Also the years are in decimal form, so it will be relabeled. There are columns like 2019 and 2020 that contain all NaNs and rows with all NaNs. They will be removed since we get no information from them. We will then get the most recent female victim rate for each country.

```python
df_intl_f = pd.read_csv('international-homicide-female.csv')

# Remove unused columns
cols_to_drop = df_intl_f.iloc[:, 1:50]
df_intl_f.drop(cols_to_drop, inplace=True, axis=1)

# Remove last two columns (2019 and 2020) since they are all NaN
cols_to_drop = df_intl_f.iloc[:, -2:]
df_intl_f.drop(cols_to_drop, inplace=True, axis=1)

# Remove rows containing metadata and set column headers
df_intl_f = df_intl_f.rename(columns=df_intl_f.iloc[3]).drop(df_intl_f.index[0:4])

# Rename column names for the years
for i in range(1, 14):
    df_intl_f.columns.values[i] = str(i+2005)

df_intl_f.rename({'Country Name': 'Country'}, axis=1, inplace=True)

# Drop rows that contain NaNs for all years
subset_f = df_intl_f.columns.tolist()[1:]
df_intl_f = df_intl_f.dropna(how='all', subset=subset_f, axis=0)

# Create a new dataframe to store the most recent numbers
df_intl = pd.DataFrame(columns=['Country', 'Female Victims'])
df_intl['Country'] = df_intl_f['Country']

# Go through the original dataset and take the most recent number for every country
for i, r in df_intl_f.iterrows():
    for val in reversed(r):
        # if the value is not NaN
        if(val==val):
            df_intl.loc[i,'Female Victims'] = val
            break
            
# Remove "World" Row
df_intl = df_intl[df_intl['Country'] != "World"]

# Sort the rate by Country        
df_intl = df_intl.sort_values(by=['Country']).reset_index(drop=True)
df_intl
```

## UN Data on Gini index
From http://data.un.org/Data.aspx?d=WDI&f=Indicator_Code%3ASI.POV.GINI

This dataset contains all the available gini index for 167 countries throughout the years. It is already quite clean. We only need to remove the footnotes. However, for some countries we have more data throughout the years than others. In order to get as many data point as possible, we decide to drop the duplicated countries while keeping the most recent data for each country. The Gini Index is a summary measure of income inequality. The Gini coefficient ranges from 0, indicating perfect equality (where everyone receives an equal share), to 1, perfect inequality (where only one recipient or group of recipients receives all the income)

```python
#import raw csv data
df_gini = pd.read_csv('combined countries gini index.csv')

#drop the last few rows of text descriptions, whose value are NaN
df_gini = df_gini.dropna()

#For each country, we want to get its gini index from the most recent years available
countries = df_gini['Country or Area'].unique()
df_gini = df_gini.sort_values(by = ['Year'])
df_gini['Year'] = pd.to_numeric(df_gini['Year'])
# df_gini = df_gini[df_gini['Year'] > 2010]
df_gini = df_gini.sort_values(by = ['Year']).reset_index()

# Drop unnecessary column "value foodnotes"
df_gini = df_gini.drop(['Value Footnotes'], axis = 1)
#Change column name
df_gini.columns = ['index', 'Country','Year', 'Value']

#Drop duplicated countries, keep the one with the most recent data
df_gini = df_gini.drop_duplicates(subset=['Country'], keep='last')

# Extract Index and Country
df_gini = df_gini.rename(columns={'Value': 'Gini'})
df_gini = df_gini[['Country', 'Gini']].sort_values(by=['Country']).reset_index(drop=True)
df_gini
```

## Human Development Index Dataset
From http://hdr.undp.org/en/indicators/137506#

The dataset is relatively clean. The data contains the value of the metadata and NaN, so the previous columns（index 0-3） of titles，rows and value are deleted. Then, set the first row to the title, so the country and HDI are set as column names for easy reference. The value range is naturalized between indices of 0-1, so the observations are clean. In addition, some columns contain the values and rows of NaN will be deleted altogether because no information is obtained from them for clear viewing.



```python
#import raw csv data
dfHDI = df = pd.read_csv("Human_Development_Index_HDI.csv")
#drop rows that are not needed
dfHDI = dfHDI.drop([0,1,2,3])
#get the columns needed
dfHDI = dfHDI[['Unnamed: 1','Unnamed: 58']]
#drop rows that are not needed
dfHDI.drop(df.tail(17).index, inplace = True)

#rename the columns
dfHDI.rename(columns={'Unnamed: 1': 'Country','Unnamed: 58':'HDI'}, inplace=True)
dfHDI = dfHDI.drop([4])
dfHDI = dfHDI.reset_index()
dfHDI = dfHDI.drop(columns='index')
dfHDI['Country'] = dfHDI['Country'].str.strip()
dfHDI
```

## Combining the datasets

```python
# Combine the dataset using a left join on the Country
df_g = df_intl
df_g = df_g.join(df_gei.set_index('Country'), on='Country', how='left')

# Since the country names do not exactly match, we will need to iterate and find a match
# Get all rows that arent matched
no_gei = df_g[df_g['GEI'].isna()]

# Iterate through the original dataset and find a country that matches with the first four characters
# if it matches then set the GEI to the corresponding country, if no match then it will remain NaN
for i1, r1 in no_gei.iterrows():
    for i2, r2 in df_gei.iterrows():
        c1 = r1['Country'].lower()
        c2 = r2['Country'].lower()
        if c1[:4] == c2[:4]:
            df_g.at[i1,'GEI'] = df_gei.at[i2,'GEI']
            
# Drop the countries with GEIs that have not been found 
df_g = df_g.sort_values(by=['Country']).reset_index(drop=True)
df_g = df_g.dropna(subset=['GEI'])
df_g
```

```python
# Add Gini dataset into the combined dataset above with matching country names
df_g = df_g[['Country','Female Victims','GEI']]
df_g = df_g.join(df_gini.set_index('Country'), on='Country', how='left')
no_gini = df_g[df_g['Gini'].isna()]

for i1, r1 in no_gini.iterrows():
    for i2, r2 in df_gini.iterrows():
        c1 = r1['Country'].lower()
        c2 = r2['Country'].lower()
        if c1[:4] == c2[:4]:
            df_g.at[i1,'Gini'] = df_gini.at[i2,'Gini']

df_g = df_g.dropna(subset=['Gini']).reset_index(drop=True)
df_g
```

```python
# Add HDI dataset into the combined dataset above with matching country names
df_g = df_g[['Country','Female Victims','GEI','Gini']]
df_g = df_g.join(dfHDI.set_index('Country'), on='Country', how='left')

no_hdi = df_g[df_g['HDI'].isna()]

for i1, r1 in no_hdi.iterrows():
    for i2, r2 in dfHDI.iterrows():
        c1 = r1['Country'].lower().strip()
        c2 = r2['Country'].lower().strip()
        if c1[:4] == c2[:4]:
            df_g.at[i1,'HDI'] = dfHDI.at[i2,'HDI']

df_g = df_g.dropna(subset=['HDI']).reset_index(drop=True)

cols = ['Female Victims','GEI','Gini', 'HDI']
df_g[cols] = df_g[cols].apply(pd.to_numeric)

df_g
```

The combined dataset will be used for EDA and data analysis. The variables include the countries and the observations are data for each country. For each country there is female victim rate and the three indices: GEI, Gini and HDI. If a observation is NaN then the row will be removed from the dataset. ## Creating the dataset for geopandas

In this section we will transform the current datasets into a dataset that can be used for geopandas. We will map the countries to a specific polygon on the world map and then include the female victims and three indices. The variables will be the country, polygon, female victim, GEI, HDI, and Gini. The observations will be the values such that we are able to generate a geoplot for visualizations.

```python
#This function standardizes the names of the countries in order to be used with geopandas

def standardize_country(country):
    countryDict = {
        "Bosnia and Herz.":"Bosnia and Herzegovina",
        "Czechia":"Czech Republic",
        "Dominican Rep.":"Dominican Republic",
        "United States of America":"United States",
        "South Korea":"Korea",
        "Kyrgyzstan": "Kyrgyz Republic",
        "Macedonia": "North Macedonia",
        "Russia": "Russian Federation",
        "Slovakia": "Slovak Republic",
    }
    if country in countryDict:
        return countryDict[country]
    return country
```

For each dataset (GEI,HDI, GINI, etc) we make their corrosponding geopandas usable dataset

```python
#Female homocides dataset

#get the geometries for each country
world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))
df_geo = world[['name', 'geometry']].rename(columns={'name': 'Country'})
# rename counries to merge
df_geo["Country"] = df_geo["Country"].apply(standardize_country)

# merge the female victim dataframe with geometries
df_geo = df_geo.join(df_intl.set_index('Country'), on='Country', how='right')

df_geo = df_geo.dropna(subset=['geometry']).reset_index(drop=True)
df_geo = geopandas.GeoDataFrame(df_geo.copy())
# turn the numbers from string to float
df_geo['Female Victims'] = df_geo['Female Victims'].astype(float)

#Male homocides Dataset

#get the geometries for each country
df_geo_m = world[['name', 'geometry']].rename(columns={'name': 'Country'})
# rename counries to merge

df_geo_m["Country"] = df_geo_m["Country"].apply(standardize_country)
# merge the male victim dataframe with geometries
df_geo_m = df_geo_m.join(df_intl2.set_index('Country'), on='Country', how='right')

df_geo_m = df_geo_m.dropna(subset=['geometry']).reset_index(drop=True)
df_geo_m = geopandas.GeoDataFrame(df_geo_m.copy())
# turn the numbers from string to float
df_geo_m['Male Victims'] = df_geo_m['Male Victims'].astype(float)


#GEI Dataset

#get the geometries for each country
df_gei_geo = world[['name', 'geometry']].rename(columns={'name': 'Country'})
# rename counries to merge
df_gei_geo["Country"] = df_gei_geo["Country"].apply(standardize_country)
# merge the GEI victim dataframe with geometries
df_gei_geo = df_gei_geo.join(df_gei.set_index('Country'), on='Country', how='right')

df_gei_geo = df_gei_geo.dropna(subset=['geometry']).reset_index(drop=True)
df_gei_geo = geopandas.GeoDataFrame(df_gei_geo.copy())
# turn the numbers from string to float
df_gei_geo['GEI'] = df_gei_geo['GEI'].astype(float)


#GINI Dataset

#get the geometries for each country
df_gini_geo = world[['name', 'geometry']].rename(columns={'name': 'Country'})
# rename counries to merge
df_gini_geo["Country"] = df_gini_geo["Country"].apply(standardize_country)
# merge the GEI victim dataframe with geometries
df_gini_geo = df_gini_geo.join(df_gini.set_index('Country'), on='Country', how='right')

df_gini_geo = df_gini_geo.dropna(subset=['geometry']).reset_index(drop=True)
df_gini_geo = geopandas.GeoDataFrame(df_gini_geo.copy())
# turn the numbers from string to float
df_gini_geo['Gini'] = df_gini_geo['Gini'].astype(float)


#HDI Dataset

#get the geometries for each country
df_hdi_geo = world[['name', 'geometry']].rename(columns={'name': 'Country'})
# rename counries to merge
df_hdi_geo["Country"] = df_hdi_geo["Country"].apply(standardize_country)
# merge the HDI victim dataframe with geometries
df_hdi_geo = df_hdi_geo.join(dfHDI.set_index('Country'), on='Country', how='right')

df_hdi_geo = df_hdi_geo.dropna(subset=['geometry']).reset_index(drop=True)
df_hdi_geo = geopandas.GeoDataFrame(df_hdi_geo.copy())
# turn the numbers from string to float
df_hdi_geo['HDI'] = df_hdi_geo['HDI'].astype(float)
```

From the scatterplot, we can determine that there is little to no correlation between the gender equality index. Given countries with high female homicide rates, we can find that their GEI is around the mean. Also, countries with low GEI do not show high rate. From the linear model, we can conclude that there is no relationship between GEI and female homicide since the slope is near horizontal. 

```python
df_gei = df[['Female Victims','GEI']]
df_gei.columns = ['fem_victims','gei']
o, p = patsy.dmatrices('fem_victims ~ gei', df_gei)
mod_log = sm.OLS(o,p)
res = mod_log.fit()
print(res.summary())
```

From the graph and linear regression model, it is very clear that there is no significant relationship between the amount of female homicide victims and the gender equality index. Our hypothesis with this variable was that there would have been some slight correlation. However the p-value is 0.9 and the R-squared value of 0 shows that there is no statistical relationship. ## Multilinear Regression Model

```python
df_cp = df.copy()
df_cp.columns = ['country','fem_victims', 'gei', 'gini','hdi']
o, p = patsy.dmatrices('fem_victims ~ gei+gini+hdi', df_cp)
mod_log = sm.OLS(o,p)
res = mod_log.fit()
print(res.summary())

#Print p-value gini index
print("\np-value for gini: ", res.pvalues[2])
```

After running a multiple regression model with the three indices together, we can see that the coefficient and the P value for each variable. With a P value of almost 0, we can see the Gini index affect the number of female homicide victims the most. An increase of 1 to the gini index means an increase of female homicide victims by 0.1376 per 100,000 females with a standard error of 0.030.

On the other hand, with a P value of 0.013, the Human Development index also plays a considerable part. An increase of 1 to the HDI means an decrease of female homicide victims by 0.0603 per 100,000 females with a standard error of 0.024.
