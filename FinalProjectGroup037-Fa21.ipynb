## Background & Prior Work

For this project, we would look into how different factors might have correlated with car accidents. For example, natural factors like weather, temperature, location might have impacts on the happening of traffic accidents and their severity. 


There are many studies and articles related to traffic accidents and to our projects. From the article “6 Ways Warmer Weather Increases Your Risk of a Car Accident” from Burrow & Associates, LLC, it is believed that `warmer weather` can lead to more traffic accidents (2015). Not only is there sun glare during sunny days, but also more people choose to dine out or cycling in hot weather. During summer time, there will be more kids and teen drivers on the road which leads to more traffic accidents. 


`Weather conditions` play an important role in car accidents. Bad weather can reduce the `visibility` of the environment for drivers and cars can be more difficult to maneuver in rain and snow.  “Driving In Bad Weather: A Risky Affair” from Safety Team suggests that bad weather can lead to car accidents and the data from the US Department of Transportation’s Federal Highway Administration (FHWA) states that there are over 5,870,000 vehicle crashes each year. 23 percent of those crashes (almost 1,312,000) are weather-related (SaferAmerica, 2015). Study “Weather as a Road Safety Hazard in Malaysia—An Overview” from Zulhaidi Mohd Jawi has similar views on the correlation between car accidents and weather conditions.Weather, particularly poor weather, is one of the dangerous operation issues that can compromise quality in all elements of road transportation, increasing the likelihood of road accidents and casualties (Jawi, 2009).


Besides reasons mentioned above, `vehicle types` can be associated with car accidents as well. In the study “Investigation of road accident severity per vehicle type” by George Yannis, trucks have higher rates of severe injuries and even fatalities, whereas buses have the highest rate of slight injuries(Yannis et al., 2017).


The `time of the day` can also have an effect on car accidents. In the article by Forbes, "Most Dangerous Times To Drive", the author Hannah Elliott concludes that there are more accidents that happen in August and Saturday is the most risky day in a week (Elliott, 2009).

## Improved Models

```python
## Choose desired categorical columns and quantitative features
cat_col = ['Weather_Condition']
quant_col = ['Distance(mi)', 'Temperature(F)', 'Wind_Chill(F)',\
             'Humidity(%)', 'Pressure(in)']
## Start with a small data sample 
train_X = df.loc[:, (cat_col+quant_col)].sample(50000, random_state = 100).dropna(axis=0)
train_Y = df.Severity.loc[train_X.index]

test_X = df.loc[:, (cat_col+quant_col)].sample(10000, random_state = 10086).dropna(axis=0)
test_Y = df.Severity.loc[test_X.index]
```

```python
num_transformer = Pipeline(steps=[ ('scaler', pp.StandardScaler()) ])
# output from Ordinal becomes input to OneHot
cat_transformer = Pipeline(steps=[('onehot', pp.OneHotEncoder(handle_unknown='ignore'))])
# preprocessing pipeline (put them together)
preproc = ColumnTransformer(transformers=[('num', num_transformer, quant_col), ('cat', cat_transformer, cat_col)])
```

### Model 1: DecisionTreeClassifier

```python
t1_start = process_time()

pl = Pipeline(steps=[('preprocessor', preproc), ('regressor', DecisionTreeClassifier())])
pl.fit(train_X, train_Y)
preds = pl.predict(test_X)
metrics.precision_score(test_Y, preds, average='micro')

t1_stop = process_time() # To track the stop time
print("Elapsed time:", t1_stop, t1_start)  
print("Elapsed time during the whole program in seconds:", t1_stop-t1_start)
metrics.precision_score(test_Y, preds, average='micro')
```

### Model 2: KNeighborsClassifier

```python
t1_start = process_time()

pl = Pipeline(steps=[('preprocessor', preproc), ('regressor', KNeighborsClassifier())])
pl.fit(train_X, train_Y)
preds = pl.predict(test_X)
metrics.precision_score(test_Y, preds, average='micro')

t1_stop = process_time() # To track the stop time
print("Elapsed time:", t1_stop, t1_start)  
print("Elapsed time during the whole program in seconds:", t1_stop-t1_start)
metrics.precision_score(test_Y, preds, average='micro')
```

### Model 3: RandomForestClassifier

```python
VType_list = ['PASSENGER', 'SPORT UTILITY VEHICLE (SUV)', 'VAN/MINI-VAN', 'PICKUP', 'TRUCK - SINGLE UNIT']
crash_features = crash_features[ crash_features.VEHICLE_TYPE.isin(VType_list) ]
features = crash_features[['MAKE', 'VEHICLE_TYPE']]
```

```python
X = pd.get_dummies(data=features, drop_first=True)
X = X.assign(DAMAGE = crash_features.DAMAGE)
```

```python
X.columns = ['MAKE_DODGE', 'MAKE_FORD', 'MAKE_HONDA', 'MAKE_HYUNDAI', \
             'MAKE_JEEP','MAKE_NISSAN', 'MAKE_TOYOTA', 'VEHICLE_TYPE_PICKUP',\
             'VEHICLE_TYPE_SUV','VEHICLE_TYPE_TRUCK', 'VEHICLE_TYPE_VAN','DAMAGE']
```

```python
# Statmodels & patsy
import patsy
import statsmodels.api as sm

# Specify our model matrix using `pastsy`.
outcome, predictors = patsy.dmatrices('DAMAGE ~ MAKE_DODGE + MAKE_FORD + MAKE_HONDA + MAKE_HYUNDAI + MAKE_JEEP + MAKE_NISSAN + MAKE_TOYOTA + VEHICLE_TYPE_PICKUP +VEHICLE_TYPE_SUV + VEHICLE_TYPE_TRUCK + VEHICLE_TYPE_VAN', X)
model = sm.OLS(outcome, predictors)
results = model.fit()
print(results.summary())
```

We conducted a __`linear probability`__ model above, with y value as a binary Damage (0 or 1), and Dummies for features like typical car make and vehicle type. 

We observe that, averagely, Car make `JEEP` and `VEHICLE_TYPE_PICKUP` are positively related to more damage; i.e. driving a JEEP car would averagely increase `1.15` percentage points possibility of higher damage. The two coefficients are both statistically significant. The underlying reason could be that the special usage of these cars lead to a higher damage possibility.

We also observe that the coefficients of car makes are pretty small, indicating that car makes are NOT statistically great factors of traffic Severity.
