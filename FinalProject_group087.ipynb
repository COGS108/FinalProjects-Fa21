# Setup

```python
# imports
import pandas as pd
import numpy as np

import seaborn as sns
sns.set(font_scale=2, style="white")

# import matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.style as style
# set plotting size parameter
plt.rcParams['figure.figsize'] = (17, 7)

# Statmodels & patsy
import patsy
import statsmodels.api as sm

# dataframes for internet users and 
# the various gdp components for a multitude of countries
df_netusers = pd.read_csv('data/countries_netusers.csv')
df_gdp = pd.read_csv('data/countries_gdp.csv')
df_govt_fce = pd.read_csv('data/govt_fce.csv')
df_hh_fce = pd.read_csv('data/household_fce.csv')
df_imports = pd.read_csv('data/imports.csv')
df_exports = pd.read_csv('data/exports.csv')
```

# Data Cleaning

Describe your data cleaning steps here.

As of now, the datasets look similar to the ones below.

To start, we can see in the second dataset shown above, there are some irrelevant columns such as 'Country Code' and 'Indicator Name' in a few of the datasets. We will drop these as we only need to focus on the 'Country Name' and year columns for our analysis. 

```python
# drop the unnecessary columns
df_govt_fce = df_govt_fce.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
df_hh_fce = df_hh_fce.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
df_imports = df_imports.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
df_exports = df_exports.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1)
```

Now all the dataframes have the same format of 'Country Name' and the years '1960' to '2020'. The issue now is that there are a lot of NA values within the datasets, especially in the early years. This makes sense as the data was probably not being recorded in some countries during those years. There are also a decent amount of NA values in recent years such as '2020' which is most likely due to the fact that the countries haven't updated their statistics.

To solve this issue we decided to only use the years '2000' to '2015', we assumed that this period in time would have the least amount of NA values and we would be able to retain enough data to use in our analysis. After isolating our desired years, we would then remove any country that had NA values, as we don't want keep any countries that have gaps in their data. 

```python
# The columns we want to keep in our datasets
# ['Country Name', '2000', ... '2016']
columns = ['Country Name']
for year in range(2000, 2016):
    columns.append(str(year))

# Transform the dataframes to have our desired columns
df_netusers = df_netusers[columns]
df_gdp = df_gdp[columns]
df_govt_fce = df_govt_fce[columns]
df_hh_fce = df_hh_fce[columns]
df_imports = df_imports[columns]
df_exports = df_exports[columns]

# Drop any rows with NA values
df_netusers = df_netusers.dropna().reset_index(drop=True)
df_gdp = df_gdp.dropna().reset_index(drop=True)
df_govt_fce = df_govt_fce.dropna().reset_index(drop=True)
df_hh_fce = df_hh_fce.dropna().reset_index(drop=True)
df_imports = df_imports.dropna().reset_index(drop=True)
df_exports = df_exports.dropna().reset_index(drop=True)
```

A problem that arises by dropping the data this way is that the datasets become mismatched and may not contain the same countries as one another. So we need to remove all the countries that are not within each dataset.

```python
# matches is a list of all the countries that are within each dataset
matches = list(set(df_gdp['Country Name'].values) & \
               set(df_netusers['Country Name'].values) & \
               set(df_govt_fce['Country Name'].values) & \
               set(df_hh_fce['Country Name'].values) & \
               set(df_imports['Country Name'].values) & \
               set(df_exports['Country Name'].values))

# Restricting datasets to only the countries within matches
df_gdp = df_gdp.loc[df_gdp['Country Name'].isin(matches)].reset_index(drop=True)
df_netusers = df_netusers.loc[df_netusers['Country Name'].isin(matches)].reset_index(drop=True)
df_govt_fce = df_govt_fce.loc[df_govt_fce['Country Name'].isin(matches)].reset_index(drop=True)
df_hh_fce = df_hh_fce.loc[df_hh_fce['Country Name'].isin(matches)].reset_index(drop=True)
df_imports = df_imports.loc[df_imports['Country Name'].isin(matches)].reset_index(drop=True)
df_exports = df_exports.loc[df_exports['Country Name'].isin(matches)].reset_index(drop=True)
```

This leaves us with 176 rows (countries) which is more than enough for our data analysis. 

To make it easier, we will use exports and imports, and combine them together into one column called net_exports.

```python
# Drop 'Country Name' column to be able to perform transformation
df_imports_v = df_imports.drop(['Country Name'], axis=1)
df_exports_v = df_exports.drop(['Country Name'], axis=1)

df_net_exports = df_exports_v.sub(df_imports_v)
df_net_exports.insert(0, 'Country Name', df_imports['Country Name'])
df_net_exports.head()
```

Now we need to combine all the datasets together. We can't do this right away as the columns in our datasets have years '2000' to '2015' as headers. What we'll need to do is utilize the melt function and transform the years into a single column 'Years'.

```python
df_netusers_melt = df_netusers.melt(id_vars = ['Country Name'], var_name = 'Year', value_name = 'internet user %')
df_gdp_melt = df_gdp.melt(id_vars = ['Country Name'], var_name = 'Year', value_name = 'GDP per capita')
df_govt_fce_melt = df_govt_fce.melt(id_vars = ['Country Name'], var_name = 'Year', value_name = 'govt_fce')
df_hh_fce_melt = df_hh_fce.melt(id_vars = ['Country Name'], var_name = 'Year', value_name = 'hh_fce')
df_net_exports_melt = df_net_exports.melt(id_vars = ['Country Name'], var_name = 'Year', value_name = 'net_exports')

df_netusers_melt
```

We have some negative observation of net exports here. We will just add the most minimum value to all values.

```python
min_net_export = df_net_exports_melt['net_exports'].min()
df_net_exports_melt['net_exports'] = df_net_exports_melt['net_exports'] - min_net_export
```

Finally, we will join all the data frames together into one 'df'. We will also change the column names into simpler ones.

```python
# join into df called df_net_gdp
df = df_netusers_melt.join(df_gdp_melt.set_index(['Country Name', 'Year']), on=['Country Name','Year'])
df = df.join(df_govt_fce_melt.set_index(['Country Name', 'Year']), on=['Country Name','Year'])
df = df.join(df_hh_fce_melt.set_index(['Country Name', 'Year']), on=['Country Name','Year'])
df = df.join(df_net_exports_melt.set_index(['Country Name', 'Year']), on=['Country Name','Year'])

# simpler column names
df.columns = ['country','year','net_percent','gdp_per_cap', 'govt_fce', 'hh_fce', 'net_exports']
df.head()
```

# Data Analysis & Results

Because we have govt_fce, hh_fce, net_export in percentage of gdp, we need to change them into current dollars per capita. Otherwise, we cannot compare between countries that have different gdp.

```python
# get percentages
df['govt_fce'] = df['govt_fce'].multiply(0.01)
df['hh_fce'] = df['hh_fce'].multiply(0.01)
df['net_exports'] = df['net_exports'].multiply(0.01)

# multiply by gdp
df['govt_fce'] = df['govt_fce'].multiply(df['gdp_per_cap'])
df['hh_fce'] = df['hh_fce'].multiply(df['gdp_per_cap'])
df['net_exports'] = df['net_exports'].multiply(df['gdp_per_cap'])

df.head()
```

### Distribution and Regression of GDP per capita

We will proceed by observing how the data looks.

```python
# plotting to see the distribution of the variables
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
sns.histplot(df['net_percent'], ax=ax1, bins=40, color='m')
sns.histplot(df['gdp_per_cap'], ax=ax2, bins=40, color='b')
ax1.set_xlabel('Internet %')
ax2.set_xlabel('GDP per capita')

sns.lmplot(
    x='net_percent',
    y='gdp_per_cap',
    data=df,
    fit_reg = False,
    aspect=1.8,
    height=10,
    scatter_kws={'color':'red'},
    line_kws={'color':'blue'}
)
```

Both the data are right skewed and their scatterplot shows heteroskedasticity so we will apply transformations to normalize them or decrease their skewness as best as we can so the assumptions of linear regression can be satisfied.

In the case of GDP per capita, a log transformation will work best. We will offset by 0.1.

For the internet users %, we will apply a square-root transformation.

```python
# Sqrt transform to normalize net percent
# log transform with a 0.1 offset to normalize gdp per cap

df['sqrt_net_percent'] = np.sqrt(df['net_percent'])
df['log_gdp_per_cap'] = np.log10(df['gdp_per_cap'] + 0.1)

# plotting to see the distribution of the variables
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True)
sns.histplot(df['sqrt_net_percent'], ax=ax1, bins=40, color='m')
sns.histplot(df['log_gdp_per_cap'], ax=ax2, bins=40, color='b')
ax1.set_xlabel('Internet %')
ax2.set_xlabel('GDP per capita')
```

The histograms above show the distribution of the data after we attempted to normalize it. They are more symmetric, less skewed, and the histogram of GDP per capita looks normal. 

The following scatter plot shows a better satisfaction of linear regression conditions and we perform linear regression on these transformed variables.

```python
# plot relationship between the net percent and GDP per capita
sns.lmplot(
    x='sqrt_net_percent',
    y='log_gdp_per_cap',
    data=df,
    fit_reg=True,
    aspect=1.8,
    height=10,
    scatter_kws={'color':'red'},
    line_kws={'color':'blue'}
)

# carry out linear regression

outcome, predictors = patsy.dmatrices('log_gdp_per_cap ~ sqrt_net_percent', df)

mod_log = sm.OLS(outcome, predictors)
fit_res = mod_log.fit()

print(fit_res.summary())
```

The scatterplot with transformed variables shows a linear relationship and roughly homoscedastic. Internet users percentage of the total population has a positive correlation with GDP per capita of the country. For each unit increase in the square root of internet user percentage, there is a statistically significant 0.2177 increase in log base 10 of GDP per capita. With a p value < 0.001 and an R squared of 0.785, we can tell there is a significance in the present of internet users in a country and the GDP per capita. 

The model is: $ \log (\text{ GDP per capita }) = 409.5822 \sqrt(\text{ internet user in percentage }) - 4346.5117$

### Distribution and Regression on Expenditures and Net exports

perform analyses on the government final consumption expenditure, household final consumption expenditure, and net exports to see which are the factors that are influenced by the internet users of a country. First, lets create histograms for all the new variables to see their distribution.

```python
# plotting to see the distribution of the variables
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=True)

sns.histplot(df['govt_fce'], ax=ax1, bins=40, color='g')
sns.histplot(df['hh_fce'], ax=ax2, bins=40, color='r')
sns.histplot(df['net_exports'], ax=ax3, bins=40, color='b')
ax1.set_xlabel('GFC Expenditure')
ax2.set_xlabel('HHFC expenditure')
ax3.set_xlabel('Net exports')
```

Both expenditures and net exports are skewed right. 

In order to apply linear regression, we will normalize them using a log transformation with an offset of 0.1.

```python
# Log transformations on new variables
df['log_govt_fce'] = np.log10(df['govt_fce'] + 0.1)
df['log_hh_fce'] = np.log10(df['hh_fce'] + 0.1)
df['log_net_exports'] = np.log10(df['net_exports'] + 0.1)

# plotting to see the distribution of the normalized variables
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=True)

sns.histplot(df['log_govt_fce'], ax=ax1, bins=40, color='g')
sns.histplot(df['log_hh_fce'], ax=ax2, bins=40, color='r')
sns.histplot(df['log_net_exports'], ax=ax3, bins=40, color='b')
ax1.set_xlabel('Log GFC Expenditure')
ax2.set_xlabel('Log HHFC expenditure')
ax3.set_xlabel('Log Net Export')
```

The histograms above show the normalized GFC Expenditure, HHFC Expenditure, and Net Exports data after we applied a log transformation. They are less skewed and unimodal.

```python
# three scatter plots so that it will be net percent versus each new group of data.
sns.lmplot(
    x='sqrt_net_percent',
    y='log_govt_fce',
    data=df,
    fit_reg=True,
    aspect=1.8,
    height=10,
    scatter_kws={'color':'red'},
    line_kws={'color':'blue'}
)

sns.lmplot(
    x='sqrt_net_percent',
    y='log_hh_fce',
    data=df,
    fit_reg=True,
    aspect=1.8,
    height=10,
    scatter_kws={'color':'red'},
    line_kws={'color':'blue'}
)

sns.lmplot(
    x='sqrt_net_percent',
    y='net_exports',
    data=df,
    fit_reg=True,
    aspect=1.8,
    height=10,
    scatter_kws={'color':'red'},
    line_kws={'color':'blue'}
)
```

The above scatterplots are created based on the GFC Expenditure, HHFC Expenditure, and the Net Exports and their relationship with net percent. 

For both expenditures, the scatter plot shows a similar shape with the scatterplot of log gdp per capita. They are fairly linear and fit well with the linear regression line.

However, the net export scatterplot still shows heteroscedasticity and nonlinear relationship with square rooted internet percentage after transformation.

```python
# carry out linear regression on log government expenditure from square rooted internet user percentage
df_govt_reg = df[['log_govt_fce', 'sqrt_net_percent']]
df_govt_reg.columns = ['govt_fce', 'net_percent']
outcome, predictors = patsy.dmatrices('govt_fce ~ net_percent', df)

mod_log = sm.OLS(outcome, predictors)
fit_res = mod_log.fit()

print(fit_res.summary())

# carry out linear regression on log household expenditure from square rooted internet user percentage
df_hh_reg = df[['log_hh_fce', 'sqrt_net_percent']]
df_hh_reg.columns = ['hh_fce', 'net_percent']
outcome, predictors = patsy.dmatrices('hh_fce ~ net_percent', df)

mod_log = sm.OLS(outcome, predictors)
fit_res = mod_log.fit()

print(fit_res.summary())

# carry out linear regression on log net export from square rooted internet user percentage
df_ne_reg = df[['net_exports', 'sqrt_net_percent']]
df_ne_reg.columns = ['net_exports', 'net_percent']
outcome, predictors = patsy.dmatrices('net_exports ~ net_percent', df)

mod_log = sm.OLS(outcome, predictors)
fit_res = mod_log.fit()

print(fit_res.summary())
```

For both expenditures, the R-squared values are between 0.7 and 0.6. This means that about 60% of the observed variation can be explained by the net percent. 

All of them has a p-value less than 5%.

Overall, our models with statistically significant coefficients are 

$ \log (\text{ household expenditure per capita }) = 409.5822 \sqrt(\text{ internet user in percentage }) - 4346.5117$

$ \log(\text{ government expenditure per capita }) = 254.4444 \sqrt(\text{ internet user in percentage }) - 735.1328 $

$ \log(\text{ net export per capita }) = 0.1294 \sqrt(\text{ internet user in percentage }) + 95.6334$
