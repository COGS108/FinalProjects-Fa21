# Overview

Our project intended to determine the relationship between (American) political affiliation and the potential spread of COVID-19 misinformation by popular news outlets on Twitter. To test our hypothesis that far-right-affiliated news outlets were more likely to spread potential COVID-19 misinformation, we used Twitter’s API to collect posts from 13 major news outlets across the American political spectrum and compared the frequency of potential misinformation keywords (e.g. Ivermectin, “Let Them Breathe,” etc.) as well as retweets, likes, followers, and post length. ## Background & Prior Work

Misinformation has become a prominent topic in recent months, particularly after the onset of COVID-19 pandemic restrictions in March 2020 and following controversy around mask-wearing, COVID-19 vaccines, and other information related to the virus. Social media platforms have shown to be a large propagator of information, and recent events have shown how information can be easily misrepresented through this medium, such as the large amounts of misinformation surrounding the COVID-19 pandemic, including concerns over vaccines and mask-wearing. A corresponding increase in polarization and politicization within U.S. media correlates with biased reporting from particular organizations, and even the spread of misinformation. Given this timely topic, our team wants to investigate which news sources are prone to spreading false information, how often this occurs, and if this misinformation spreads to social media users and comments. 

One example of prior research that focused on finding the correlation between politicized disinformation and Twitter was Matthew Hindman and Vlad Barash’s research for the Knight Foundation entitled “Disinformation, ‘Fake News,’ and Influence Campaigns on Twitter.” The team analyzed Twitter accounts connected to “fake news” or conspiracy news outlets from the month of the 2016 US Presidential Election to five months after the election. While the focus of their research was very broad, some of the most notable findings were that approximately 89 percent of disinformation-based tweets were made by the top 50 conspiracy outlets and that more than “6.6 million tweets” related to conspiracies were spread on Twitter the month before the 2016 election. Hindman and Barash’s research focus is similar to our proposed research, so we can apply their methods of tracking Twitter post keywords and tagging posts as “misinformation” within our own project. 

Also relevant to our proposed research is “Polarization, Partisanship and Junk News Consumption on Social Media During the 2018 US Midterm Elections,” conducted at Oxford University by Nahema Marchal, Lisa-Maria Neudert, Bence Kollanyi, and Philip N. Howard. The researchers studied “2.5 million tweets and 6,986 Facebook pages over a 30-day period” before the 2018 US midterms to determine sources and primary audiences of what they call “junk news.” The team found that even more “junk news” was circulated in 2018 than during the 2016 US presidential election, that most of the sources referenced were not from reputable sources (e.g. “public agencies, experts or the political candidates themselves”), and that the audience of “junk news” had expanded to include “mainstream conservative social media users.” The team’s discovery that instances of misinformation multiplied between 2016 and 2018 suggests that misinformation may be even more common in 2019-2021. Though this research topic varies from our own focus on COVID-19, we should note in our data analysis and discussion that based on their findings, the sharing of “official” news sources is relatively low compared to the overall sharing of sources referenced on social media.

```python
dataframes = [theintercept, jacobin, vox, msnbc, cnn, ap, abc, pbs, wsj, nypost, fox, newsmax, oann]
for x in dataframes:
    x["text"] = x["text"].str.lower()
    x["length_text"] = x["text"].str.len()
```

```python
def num_minsinfo(df):
    def count_misinfo(text):
        count = 0
        keywords = ["ivermectin", "bioterrorism", "hydroxychloroquine", "anti-lockdown", "crystals", "hoax",
                    "bio-weapon", "big pharma", "government control", "forced control", "forced compliance",
                    "just the flu", "medical tyranny", "let them breathe", "resist the muzzle", "masks kill",
                    "let kids breathe", "kids need oxygen", "test subjects", "5g", "medical apartheid", "vaccine shedding"]
        for x in keywords:
            if x in text:
                count += 1
        return count
    return df["text"].apply(count_misinfo)
for x in dataframes:
    x["num_misinfo_words"] = num_minsinfo(x)
misinfo = [num_minsinfo(x).sum() for x in dataframes]
misinfo
```

*Subset Question 1:*

Through Figure 1, we found that far-right news outlet OANN, right-wing news outlets NYPost and Fox, and far-left news outlets like Jacobin and TheIntercept generally had greater frequencies of misinformation keywords in their posts. The inclusion of the far-left sources contradicts our hypothesis since we assumed that only far right-wing news outlets would have the highest misinformation keyword frequency. Instead, the magnitude of political skew seems to be the deciding factor: the farther left or right a news source is, the more likely they are to register in our findings. It is possible that the keywords the left-wing sources used were strictly for reporting misinformation and, as a result, does not indicate use of misinformation. Another possibility could be that these news outlets, regardless of political affiliation, are in fact perpetuating COVID-19 misinformation. The Ad Fontes Media Bias Chart® suggests this possibility, as its news outlets are arranged in a horseshoe shape, with news sources on the extremes being more susceptible to general misinformation, though further testing would be needed to see if this generalizes to COVID-19 misinformation.


*Subset Question 2:*

Based on Figures 2-14, we have found that far-right news outlets like OANN, right wing-news outlets NYPost and Fox, and center news outlets like AP had the most-liked tweets. If we eliminate news outlets that do not have a high follower count and a high keyword frequency, then the news outlets that could be spreading the most misinformation would be OANN, Fox (both with a 1.4 million follower count) and NYPost (with a 2.3 million follower count), which partly aligns with our hypothesis. A possible scenario based on Figure 15-27 could be that the news outlets’ posts that have a higher length of text could also be generating the most likes and not because people are agreeing with the misinformation. Also, based on Figure 28-40, news outlets’ posts that are being retweeted primarily do not contain our COVID-19 misinformation keywords, suggesting that this measure is not relevant to our research question. As a result, this method of analysing the interactions with posts does not completely correlate with the spread of COVID-19 misinformation. Additionally, people who are trying to fact check misinformation could also be liking posts made by these news outlets to spread and show approval for factual information. Of course, there is the possibility that people are interacting with these posts to either spread misinformation or to contradict the facts that the news outlets are sharing.

# Ethics & Privacy

There are a range of potential issues that we encountered in regard to biases and privacy with the data we collected. For example, by only collecting data from certain news stations using certain keywords, we risk missing prevalent keyword usage found in other stations or with other keywords. Headlines in relation to selected keywords may be misleading or be considered clickbait about the content of our questions. In terms of privacy, we processed Tweets made by users or organizations potentially unaware of data collection through Twitter’s API accessibility despite EULAs. However, we have removed any usernames attached to the tweets and only kept the tweet’s ID and its time of creation. This ID is only searchable through Twitter’s API, which can only be accessed by manually-approved developer accounts. Thus, tweets made by users can remain anonymous, except for those by the news stations we analyzed. We also considered the fact that by selecting only a few English-speaking, large news stations, we are missing out on potentially relevant information from small or non-English-speaking news stations based in the United States, including trends from news stations catering to subgroups within our population.

# Conclusion & Discussion

To determine the relationship between the potential spread of COVID-19 misinformation by news outlets and their political affiliation (compared with other news outlets across the American political spectrum), we collected information on recent Twitter posts from an array of 13 news outlets across the American political spectrum using Twitter’s API. We divided this into two subset questions. First, at what frequency did news outlets use misinformation keywords? Second, which news outlet’s posts with misinformation keywords were retweeted and favorited the most? To answer these questions, we compared Twitter post data via bar charts of relative keyword frequencies, histograms of relative popularity (via likes), and scatterplots of the word count to likes frequency for each news outlet. We also created a scatterplot for each news outlet to compare the number of retweets of misinformation-related and unrelated posts. We hypothesized that far-right news sources would both use the most misinformation keywords and have the most engagement on posts with said keywords. 
 
From our analysis, we found that our hypothesis was incorrect for our first subset question, but was correct for our second subset questions. Though far-right OANN, right-wing NYPost, and right-wing Fox did have high frequencies of misinformation keywords, center PBS and far-left news outlets Jacobin and TheIntercept also had high relative frequencies of misinformation keywords, failing our first subset hypothesis. However, when we compared follower and like counts per news outlet to the relative frequency of misinformation keywords in posts, far-right OANN, right-wing NYPost, and right-wing Fox shared the greatest potential “spread” of misinformation, which matched our hypothesis being that right-wing news outlets would have the highest misinformed post interactions.

Since our findings were inconclusive, we cannot conclude whether political affiliation of news outlets is correlated with the potential spread of COVID-19 misinformation. Our data collection methods did not account for the use of misinformation terms to report on misinformation spread, which generated false positives (as with center PBS and far-left Jacobin). We also did not account for the possibility that likes and retweets might be used for other goals than the support of misinformation (e.g., combating misinformation or trolling to encourage misinformation). Future efforts should test for false positives flagged as “spreading misinformation” (via sentiment analysis or by cross-checking with lists of “neutral” or “anti-misinformation” keywords). Repeating this research with refined data collection and analysis methods could generate useful information about the spread of misinformation over Twitter and be extended to other social media platforms. A better understanding of how and by whom misinformation about nationally significant topics are spread could help remove misinformation posts as they appear, building Americans’ overall trust in news outlets as sources for factual information.
