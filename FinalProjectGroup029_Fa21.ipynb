### Merging COVID-19 Data with Economic Matrix 2020-2021

In this section, we will merge our economic matrix with our COVID-19 data, however, we will only care about the locations we have in both datasets, therefore we will use inner join to prepare the data for further correlation, comparing analysis.

As we mentioned above, we will leave the empty value NaN as design, which we will only analysis specific indecators or COVID-19 data one at a time.

Additionally, we will only preserve the maximum value of covid data in each mon per country, given that we only have one measurement of economic indicators for each month.

```python
covid_combine_month = covid_combine.copy()
covid_combine_month['DATE'] = pd.DatetimeIndex(covid_combine_month['DATE'])
covid_combine_month['MONTH'] = covid_combine_month['DATE'].dt.to_period('M')
covid_combine_month.head()
```

```python
country_month_data = covid_combine_month[['COUNTRY', 'DATE', 'MONTH', 'Confirmed', 'Deaths',
       'stringency_index', 'population', 'gdp_per_capita',
       'human_development_index']].groupby(['COUNTRY', 'MONTH']).max().reset_index()
country_month_data['DATE'] = country_month_data['DATE'].dt.to_period('M').dt.to_timestamp()
country_region = covid_combine_month.groupby(['COUNTRY', 'WHO Region']).count().reset_index()[['COUNTRY', 'WHO Region']]
country_location = covid_combine_month.groupby(['COUNTRY', 'LOCATION']).count().reset_index()[['COUNTRY', 'LOCATION']]
df_covid_month = country_month_data.merge(country_location, on='COUNTRY', how='left')
df_covid_month = df_covid_month.merge(country_region, on='COUNTRY', how='left')

df_covid_month.head()
```

```python
def month_to_dt(month):
  #convert month in the covid dataset to a datetime object
    if month < 12:
        return datetime(2020, month, 1)
    else:
        return datetime(2019, month, 1)

def convert_date(date):
  #convert date string in the economy dataset to datetime object
    return datetime.strptime(date, "%Y-%m")
```

```python
df_economy = df_Economic.copy()
df_economy['DATE'] = df_economy['DATE'].apply(convert_date)
df_economy = df_economy[(df_economy['DATE']>datetime(2019, 11, 1)) & (df_economy['DATE']<datetime(2020,11,1))]
df_economy['DATE'] = pd.to_datetime(df_economy['DATE'])
df_economy = df_economy[['LOCATION', 'COUNTRY', 'DATE', 'CLI', 'CPI_GY', 'CPI_GP', 'GDP', 'Net_Trade_USD_Billions', 'Unemployment']]
df_economy.head()
```

```python
econ_covid = pd.merge(df_covid_month, df_economy, on = ['COUNTRY', 'DATE'], how = 'inner').drop(['LOCATION_x', 'LOCATION_y'], axis = 1)
econ_covid.head()
```

### Fill the NaN in Population using Population 2020

```python
population_df.rename(columns = {'Country (or dependency)':'Country', 'Population (2020)':'Population'}, inplace = True)
population_df = population_df.replace({'Czech Republic (Czechia)' : 'Czech Republic'})
```

```python
def get_population(country, population_df):
  try:
    population = population_df[population_df['Country'] == country]['Population'].values[0]
  except:
    return np.nan
  return population
```

### Fill the NaN in gdp_per_capita and human_development_index

```python
econ_covid = econ_covid.drop('population', axis = 1)
econ_covid['Population'] = econ_covid['COUNTRY'].apply(get_population, population_df = population_df)
econ_covid.Population.isna().sum()
```

```python
econ_covid['gdp_per_capita'] = econ_covid['gdp_per_capita'].fillna(econ_covid.groupby('COUNTRY')['gdp_per_capita'].transform('mean'))
econ_covid['human_development_index'] = econ_covid['human_development_index'].fillna(econ_covid.groupby('COUNTRY')['human_development_index'].transform('mean'))
econ_covid.head()
```

### Standardize Cases and Deaths
Given that the number of cases and deaths for each countries are absolute number, we need to transform them into standardized number by dividing them with its population.

```python
econ_covid['Confirmed_rate'] = econ_covid['Confirmed']/econ_covid['Population']
econ_covid['Deaths_rate'] = econ_covid['Deaths']/econ_covid['Population']
```

## Segmenting countries into groups

The countries are divided into 3 groups based on Human Development Index(HDI): 
- Country with Low HDI: HDI < 0.85
- Country with Intermediate HDI: 0.85 <= HDI < 0.9
- Country with High HDI: HDI >= 0.9

```python
#Separate the countries into 3 groups based on human development index
country_HDI = econ_covid.groupby('COUNTRY').mean().reset_index()[['COUNTRY', 'human_development_index']]
low_HDI_country = country_HDI[country_HDI['human_development_index']<0.85]['COUNTRY'].to_list()
mid_HDI_country = country_HDI[(country_HDI['human_development_index']<0.9) & (country_HDI['human_development_index']>=0.85)]['COUNTRY'].to_list()
high_HDI_country = country_HDI[country_HDI['human_development_index']>= 0.9]['COUNTRY'].to_list()
```

## How are difference among the impact affected by the pandemic on groups of countries? What will the significance be?

The target difference we will be discuss is the *impact* under pandemic, therefore, we will do experiments as well as statistc tests on the difference (decline/rise) of each indicators during the pandemic, then we will test the differences of these values between  different groups of countries.

The criteria we will be using to calculate differences:

Subtract each country’s beginning month with minimum/maximum value 
    - some value will take minimum, for example GDP since GDP tends to decline during the pandemic, the others will take maximum, for example Unemployment since Unemployment rate tends to increase during the pandemic. 


```python
#give each country a label of low, mid, high
def get_HDI_label(country, group_low, group_mid, group_high):
    if country in group_low:
        return 'Low'
    elif country in group_mid:
        return 'Mid'
    else:
        return 'High'

#Calculate the difference of the economic indicators over time of each country
def get_indicator_difference(econ_covid_df, indicator, mode='min'):
    #use max for "Unemployment" and min for other indicators
    if mode == 'min':
        countries = econ_covid_df['COUNTRY'].unique()
        df = pd.DataFrame(columns=['COUNTRY', 'first_month', 'last_month', 'min_month','diff_last_first', 'diff_min_first'])
        for country in countries:
            country_df = econ_covid_df[econ_covid_df['COUNTRY'] == country].reset_index()
            first_month = country_df.iloc[0]['DATE']
            first_month_stat = country_df.iloc[0][indicator]
            last_month = country_df.iloc[-1]['DATE']
            last_month_stat = country_df.iloc[-1][indicator]
            min_month_idx = country_df[indicator].idxmin()
            if math.isnan(min_month_idx):
                min_month = np.nan
                min_month_stat = np.nan
            else:
                min_month = country_df.iloc[min_month_idx]['DATE']
                min_month_stat = country_df.iloc[min_month_idx][indicator]
            try:
                diff_last_first = last_month_stat - first_month_stat
            except:
                diff_last_first = np.nan

            try:
                diff_min_first = min_month_stat - first_month_stat
            except:
                diff_last_first = np.nan

            df = df.append({'COUNTRY':country, 'first_month':first_month, 'last_month':last_month, 'min_month':min_month, 
                           'diff_last_first': diff_last_first, 'diff_min_first': diff_min_first}, ignore_index = True)
    else:
        countries = econ_covid_df['COUNTRY'].unique()
        df = pd.DataFrame(columns=['COUNTRY', 'first_month', 'last_month', 'max_month','diff_last_first', 'diff_max_first'])
        for country in countries:
            country_df = econ_covid_df[econ_covid_df['COUNTRY'] == country].reset_index()
            first_month = country_df.iloc[0]['DATE']
            first_month_stat = country_df.iloc[0][indicator]
            last_month = country_df.iloc[-1]['DATE']
            last_month_stat = country_df.iloc[-1][indicator]
            max_month_idx = country_df[indicator].idxmax()
            if math.isnan(max_month_idx):
                max_month = np.nan
                max_month_stat = np.nan
            else:
                max_month = country_df.iloc[max_month_idx]['DATE']
                max_month_stat = country_df.iloc[max_month_idx][indicator]
            try:
                diff_last_first = last_month_stat - first_month_stat
            except:
                diff_last_first = np.nan

            try:
                diff_max_first = max_month_stat - first_month_stat
            except:
                diff_last_first = np.nan

            df = df.append({'COUNTRY':country, 'first_month':first_month, 'last_month':last_month, 'max_month':max_month, 
                           'diff_last_first': diff_last_first, 'diff_max_first': diff_max_first}, ignore_index = True)
    df['HDI_label'] = df['COUNTRY'].apply(get_HDI_label, group_low=low_HDI_country, 
                                                          group_mid=mid_HDI_country, group_high=high_HDI_country)
    return df
```

### 1. Permutation Test

1. For each indicator, we do permutation test on pair of groups (Low High, Low Mid, Mid High)
2. Calculate diff as the group statistics, ex. s = mean(High) - mean(Low)
3. shuffle label of groups, calculate s' = new_mean(High) - new_mean(Low)
4. Repeat step3 2000 times, plot out the distribution histagram
5. Lable out the statistics difference on distribution histagram s, and calculate the p-value of s
6. Back to step one, do permutation test on each pair

```python
# Function for generating a permutation test on speculating whether the mean differences of a indicator of two groups of countries are similar.
def permutation_indicator_diff_between_groups(indicator, groups, N_rep):
  if indicator != 'Unemployment':
    ind_diff_df = get_indicator_difference(econ_covid, indicator, 'min').astype({'diff_last_first': 'float', 'diff_min_first': 'float'})
    obs_diff_df = ind_diff_df.groupby('HDI_label').mean().loc[:, 'diff_min_first']
    obs_diff = abs(obs_diff_df.loc[groups[0]] - obs_diff_df.loc[groups[1]])

    differences = []
    copy = ind_diff_df
    for i in range(N_rep):
      # shuffle the HDI_label
      shuffled_HDI = (
        copy['HDI_label']
        .sample(replace=False, frac=1)
        .reset_index(drop=True)
      )
    
      # put them in a table
      copy['HDI_label'] = shuffled_HDI
    
      # compute the group differences (test statistic!)
      group_means = (
        copy
        .groupby('HDI_label')
        .mean()
        .loc[:, 'diff_min_first']
      )
      difference = abs(group_means.loc[groups[0]] - group_means.loc[groups[1]])
    
      # add it to the list of results
      differences.append(difference)
    title = f'average differences in {indicator} between {groups[0]} and {groups[1]}'
    pd.Series(differences).plot(kind='hist', density=True, title=title)
    plt.scatter(obs_diff, 0.01, color='red', s=40, zorder=2);
    p_val = (differences > obs_diff).mean() 
    print(f'p-value of the permutation test on difference of {indicator} between {groups[0]} and {groups[1]} is {p_val}')

  else:
    ind_diff_df = get_indicator_difference(econ_covid, indicator, 'max').astype({'diff_last_first': 'float', 'diff_max_first': 'float'})
    obs_diff_df = ind_diff_df.groupby('HDI_label').mean().loc[:, 'diff_max_first']
    obs_diff = abs(obs_diff_df.loc[groups[0]] - obs_diff_df.loc[groups[1]])

    differences = []
    copy = ind_diff_df
    for i in range(N_rep):
      # shuffle the HDI_label
      shuffled_HDI = (
        copy['HDI_label']
        .sample(replace=False, frac=1)
        .reset_index(drop=True)
      )
    
      # put them in a table
      copy['HDI_label'] = shuffled_HDI
    
      # compute the group differences (test statistic!)
      group_means = (
        copy
        .groupby('HDI_label')
        .mean()
        .loc[:, 'diff_max_first']
      )
      difference = abs(group_means.loc[groups[0]] - group_means.loc[groups[1]])
    
      # add it to the list of results
      differences.append(difference)
    title = f'average differences in {indicator} between {groups[0]} and {groups[1]}'
    pd.Series(differences).plot(kind='hist', density=True, title=title)
    plt.scatter(obs_diff, 0.01, color='red', s=40, zorder=2);
    p_val = (differences > obs_diff).mean() 
    print(f'p-value of the permutation test on difference of {indicator} between {groups[0]} and {groups[1]} is {p_val}')
```

### 2. One sample t-test and median test against total mean
(requirement: normal population)

Do one sample t-test with group mean on High, Mid, Low groups seperately

```python
def one_sample_t_test_with_mean(group, indicator, diff_column, title, mode='min'):
    #extract the data
    stat_diff_df = get_indicator_difference(econ_covid, indicator, mode)
    mean_stat_diff = stat_diff_df[diff_column].mean()
    group_stats = stat_diff_df[stat_diff_df['HDI_label']==group][diff_column].dropna()
    #ploting the distribution and the mean
    plt.figure(figsize=(8, 4))
    sns.histplot(group_stats, bins = 15)
    plt.axvline(x=mean_stat_diff, color='red', label='mean differences of all countries')
    plt.legend()
    plt.title(title)
    #perform one sample t test
    result = stats.ttest_1samp(group_stats, mean_stat_diff)
    return result
```

### 3. Two sample t-test
(requirment: normal population)

Do two sample t-test on mean in pairs (Low High, Low Mid, Mid High)



```python
def two_sample_t_test_with_mean(group1, group2, indicator, diff_column, title, mode='min'):
    #extract the data
    stat_diff_df = get_indicator_difference(econ_covid, indicator, mode)
    df = stat_diff_df[(stat_diff_df['HDI_label']==group1) | (stat_diff_df['HDI_label']==group2)]
    df = df.dropna(subset=[diff_column])
    #plot
    plt.figure(figsize=(8, 4))
    sns.histplot(df, x=diff_column, hue='HDI_label')
    plt.title(title)
    #2 sample t test
    result = stats.ttest_ind(df[df['HDI_label']==group1][diff_column], df[df['HDI_label']==group2][diff_column])
    return result
```

### 4. ANOVA

Analysis of variance for muli-category, comparing the mean across groups
requirments: 1. Homogeneity of variance, 2. Normal, 3. independent sampling



```python
def ANOVA(indicator, diff_column, mode):
    stat_diff_df = get_indicator_difference(econ_covid, indicator, mode)
    data1 = stat_diff_df[stat_diff_df['HDI_label'] == 'Low'][diff_column].dropna()
    data2 = stat_diff_df[stat_diff_df['HDI_label'] == 'Mid'][diff_column].dropna()
    data3 = stat_diff_df[stat_diff_df['HDI_label'] == 'High'][diff_column].dropna()
    return f_oneway(data1, data2, data3)
```

```python
fvalue_1, pvalue_1 = ANOVA('CPI_GY', 'diff_min_first', 'min')
fvalue_2, pvalue_2 = ANOVA('CLI', 'diff_min_first', 'min')
fvalue_3, pvalue_3 = ANOVA('GDP', 'diff_min_first', 'min')
fvalue_4, pvalue_4 = ANOVA('Unemployment', 'diff_max_first', 'max')
fvalue_5, pvalue_5 = ANOVA('Net_Trade_USD_Billions', 'diff_min_first', 'min')
print('CPI_GY f-value: ' + str(fvalue_1) + ' p-value: ' + str(pvalue_1))
print('CLI f-value: ' + str(fvalue_2) + ' p-value: ' + str(pvalue_2))
print('GDP f-value: ' + str(fvalue_3) + ' p-value: ' + str(pvalue_3))
print('Unemployment f-value: ' + str(fvalue_4) + ' p-value: ' + str(pvalue_4))
print('Net_Trade_USD_Billions f-value: ' + str(fvalue_5) + ' p-value: ' + str(pvalue_5))
```

In general, the f-value in ANOVA analysis for each indicators are all relatively low with big p-values, therefore in multi-categorical level, there are no significant difference bwteen the change of indicators among groups

### 5. Wilcoxon Tests (Wilcoxon Rank-Sum test)
(requirement: symmetrical distribution)
Test the difference in means of the populations for different groups 
We will do the test in pairs like two sample t-test



```python
def wilcoxon_test(group1, group2, indicator, diff_column, mode):
    stat_diff_df = get_indicator_difference(econ_covid, indicator, mode)
    data1 = stat_diff_df[stat_diff_df['HDI_label'] == group1][diff_column].dropna()
    data2 = stat_diff_df[stat_diff_df['HDI_label'] == group2][diff_column].dropna()
    return ranksums(data1, data2)
```

### 6. Kruskal Wallis test(K-W Test)
(No requirement on population)
Similar to ANOVA, kruskal test can be use to measure the difference in mean across multiple groups

```python
def ks_test(indicator, diff_column, mode):
    stat_diff_df = get_indicator_difference(econ_covid, indicator, mode)
    data1 = stat_diff_df[stat_diff_df['HDI_label'] == 'Low'][diff_column].dropna().values
    data2 = stat_diff_df[stat_diff_df['HDI_label'] == 'Mid'][diff_column].dropna().values
    data3 = stat_diff_df[stat_diff_df['HDI_label'] == 'High'][diff_column].dropna().values
    return kruskal(data1, data2)
```
