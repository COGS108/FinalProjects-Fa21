## Dataset 1: Job Openings.

### How clean is the data?
  
This dataset was provided by the Bureau of Labor Statistics. The dataset was not in a clean state. There are commas in the observations, dates are not in proper formats, and the columns are not of proper int types. 


### Steps to usable format?
  
First, we had to use 'to_datetime' to standardize the 'Date' column into a proper format, which also changed the dtype to 'datetime64'. Second, most, if not all, rows (except 'Data' column) had commas. We fixed all the rows by replace the comma with "" and it is still an object type. Third, since these columns were objects (date column is okay), we needed to select an apply all object types in the columns and apply the 'to_numeric' function. 
  
  
### Necessary pre-processing steps?
  
We did not have to take pre-processing steps. All we needed to do was set it to a usable state and it is ready to be analyzed!

```python
display(df_job_openings)
display(df_job_openings.dtypes)
```

We can see that the column values are not appropriate types with commas and our date format is not good for analyzing later on. Let's starting cleaning...

```python
# Accepts column to format date
'''
Formats date
    from:   "Month YYYY"
    to:     "YYYY-MM"

'''
def standardize_date_1(df):
    df['Date'] = df['Date'].apply(pd.to_datetime)


# Accepts column to format numbers with commas
'''
Formats commas out of total count
    from:   "000,000"
    to:     "000000"

'''
def standardize_count(df):
    df.replace(',','', regex=True, inplace=True)
    types = df.select_dtypes(object).columns 
    df[types] = df[types].apply(pd.to_numeric, errors = 'coerce')
```

```python
standardize_date_1(df_job_openings)
standardize_count(df_job_openings)

df_job_openings.drop(columns=['Unnamed: 10'], inplace=True)

display(df_job_openings)
display(df_job_openings.dtypes)
```

```python
# rename columns
df_job_openings.columns = ['date','nfarm_open', 'nfarm_hire', 'nfarm_sep', 'priv_open', 'priv_hire', 'priv_sep', 'gov_open', 'gov_hire', 'gov_sep']
df_job_openings.head()
```

We have now shown that this dataset is now suitable for analysis as we have uniform entries, and the units of measurements and data types are appropriate. Overall, the dataset is clean to begin our EDA.

### Analysis
  
In our analysis we will not discriminate against any groups of people. Transparency is our goal as we intended to avoid malicious and negligent use of data that may harm any group(s).
