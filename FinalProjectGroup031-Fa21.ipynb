## Overview

The purpose of our project is to see how students’ predictions of their starting salaries relate to the actual numbers. We collected survey data from students at UCSD through social media, peer connections, and class discussion boards in order to gather data to answer this question. We asked the survey respondents several questions, including but not limited to, their major, starting salary expectations, post-graduation plans, and whether or not they had done prior research into their expected starting salary. Through these questions, we plan to see how accurate students are in regards to their salary predictions, and if these predictions change depending on their major, post-graduation plans, or research into expected salaries. To compare to our survey results we found data from the Department of Education on expected starting salaries based on major for universities across the United States (and then filtered for UCSD specifically) and from the Wall Street Journal detailing average starting salaries nationwide depending on undergraduate major. We then created several figures comparing our survey results to the national averages in order to determine how accurate UCSD students are at predicting their starting salary. Our results concluded that UCSD students do in fact overestimate their starting salaries, but that the factors we measured had no strong correlation with our results.

## Research Question

Do UCSD students overestimate their starting salaries? Does a student's undergraduate major have any influence on how close their estimate is to previously collected data? What about prior research into starting salaries, or admitted level of influence career/salary prospects had on them choosing their major?

## Background & Prior Work

As college students, this question is extremely relevant to us. We have all been told throughout our years of schooling that college is the optimal route for making a large salary and having a successful career. One study found that 17% of college students expect to make 85000 dollars or more from their first job out of college [4]. Many students have good reason to believe this, as starting salaries for the class of 2020 were more than 10% higher than the class of 2018’s [1]. We know that many students are confident in their abilities and their determination to make a large salary. Despite this, we feel it is important for students to have realistic expectations for post graduate options in order to make the decisions that are best for them. Studies done on this topic in the past have shown that there is a significant gap between expected salaries of graduating students and their actual salaries post-graduation. One study concludes that students overestimate their salaries by 23%, with business majors being the worst predictors [2]. Other studies compare expected starting salaries from students from different schools to see which schools had the most realistic students [3]. We wanted to see if this difference in expected vs starting salary would also be seen in UCSD students, and if the same majors were demonstrated to be overestimating or underestimating their salaries. The mentioned studies served as a helpful starting point to understanding the tasks we needed to complete in order to begin working on this project. The comparisons of different majors is something we focused on, with the hope of understanding which majors students overestimate or underestimate their starting salaries the most. We also wanted to include questions in our survey asking whether or not students plan to begin their career after graduation or to continue their education, since that could have an effect on their predicted and expected salaries. Finally, we also wanted to see if the amount of research a student did prior to answering this survey had any effect on their predictions, since this was not included in any of the previous studies we looked into. Ultimately, college is extremely costly. With this project, we hope that the results can help students make more informed decisions about their education and gauge if going to a university is worth it for them.

```python
doe_ratios
```

```python
ax = doe_ratios.plot(kind='barh', y='survey/doe', color='y', title='Ratio of Survey Responses to DOE Data')
ax.axvline(1, color='darkgray', linewidth=2, linestyle='dashed');
```

From this, it is fairly obvious that, when comparing against the DOE dataset, UCSD students generally provide an overestimate of their starting salaries that is up to 2.5 times the DOE-reported value.

Next, in order to determine how much the amount of research UCSD students have done regarding the starting salary of their major and its impact on their major choice affects the accuracy of their estimated starting salary, we will compare the average influence and research of each major to how accurate their starting salary estimate is. We measure the accuracy of their prediction by first examining the ratio of their prediction to the actual value according to the DOE dataset. Because the ratio of 1 shows an ideal guess, we will subtract 1 from the ratios and then take the absolute value. This way, we will see the distance the ratios are from 1, showing us how close they are to being accurate.

```python
doe_ratios = doe_ratios.reset_index().rename(columns={"index": "major"})
survey_means_ratio = pd.merge(survey_grouped_mean, doe_ratios, on="major").drop('starting_salary', axis=1).set_index('major', drop=True)
```

```python
survey_means_ratio = survey_means_ratio.rename(columns={'survey/doe' : 'doe_ratios_inaccuracy'})
survey_means_ratio['doe_ratios_inaccuracy'] = survey_means_ratio['doe_ratios_inaccuracy'].subtract(np.ones(len(survey_means_ratio['doe_ratios_inaccuracy']))).abs()
```

```python
# Compare all generated barcharts
survey_means_ratio.plot(kind= 'bar' , secondary_y= 'doe_ratios_inaccuracy' , rot= 90, figsize=(18, 6), color=['r','g','y'], title='Research and Influence Against Estimation Inaccuracy (DOE)');
```

In the graph above, we analyze whether the research and influence values affect the DOE ratio's inaccuracy values. In Civil Engineering, it can be seen that the research and influence values are high, and the ratio's inaccuracy value is low. Thus, it appears that with more research and influence, a better estimation is made. Howerver, after further analysis, it can be seen that the major Ecology, Evolution, Systematics, and Population Biology has high accuracy despite low research and influence. The rest of the graph also does not seem to follow a clear pattern, so it can be inferred that research and influence do not have an effect on the accuracy of a UCSD student's starting salary prediction.

## Ethics & Privacy

For our survey data, all information was collected anonymously and with the consent of the responders. However, there are a few things about our survey that could have potentially caused bias in our results. Firstly, it was promoted through social media platforms like Facebook, Campuswire, Reddit, and by word of mouth. This could have potentially produced bias as only those who use those platforms and those who happen to hear about it from friends would be answering the survey. Survey questions regarding prior knowledge on the topic and plans after college could influence what the responder puts down as their expected salary. The question asking about expected salary had a note to only include numeric values, and the figure we used as an example was 80000. Seeing that figure might have influenced what the responders put as their expected salary. Another important thing to consider when working with surveys is that anyone can fill it out with any information they like, so the responses cannot be fully reliable. Furthermore, the number of responses was relatively low compared to the total student population. A smaller sample size usually leads to less reliable results as there is more variance in the data.

Our second dataset containing median UCSD undergraduate starting salaries was obtained from the Department of Education. However, only those who had received federal financial aid in the form of grants or loans were recorded in the dataset. Those who graduated without any form of financial aid could skew the data significantly. Students who do not need financial aid usually come from better situations and might have access to better opportunities post-graduation.

Our other dataset is from Kaggle, and it is included in an independent report by the Washington Journal analyzing the highest paid majors out of university. This data is from over 5 years ago, so many of the figures could have changed throughout that time. Another problem is that some majors in the dataset are either not offered here at UCSD, or are under a different name. For that reason, we had to leave out some data pertaining to some majors. For these reasons, it would be irresponsible to claim that our results are fully representative of UCSD students.
