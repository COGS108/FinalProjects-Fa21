## Background & Prior Work

Healthcare has always been a controversial topic in American society due to the private healthcare system. Access to high-quality treatment often comes with a big price tag that the average American cannot afford leaving them to either decline treatment or fall into debt. This meant that American citizens weren't able to obtain the proper care for their health issues which would ultimately affect the mortality rate of Americans. 

However, under the Obama administration, the Affordable Care Act was introduced in 2010 to decrease the cost of health insurance for those that needed it. This meant that access to healthcare became a lot more affordable than it was compared to the past. This raises the question of whether an increase of access to healthcare will increase the life expectancy in the United States per person specifically looking at the data from before 2010 and comparing it with data after 2010. 

States can also expand their Medicaid program with their own funds to allow their residents a cheaper cost of health insurance. Again, we are curious whether these expansions have any impact on the life expectancy of the average American compared to those that live in states without any form of Medicaid expansion. 

Previous research has found that adequate access to healthcare is associated with lower mortality rates in some Chinese populations (Hao, Lisha, et al.), and speculative research suggests that implementing a universal healthcare system could be beneficial in countries that haven't recently seen significant growth in their life expectancies (Ranabhat, Chhabi L., et al., "Uneven Access..."). Moreover, it has been found that certain populations in the US including the near-elderly have worse health outcomes when they have inadequate access to health insurance (Woolhandler, Steffie, Himmelstein).

### State Health Insurance Coverage Data
In this dataset, we will fill in the `null` values and convert the percentages to decimal values. The first step is to read in the csv file as seen below. 

````python
# grabbing some of the state stratified health insurance rate data
state_insur = pd.read_csv('states.csv')
state_insur.head()
````

We then need to clean up all null values that may exist in this dataset. To do so, we can find the number of null values that exist per column in the state_insur dataset. If the number of null values is greater than 0, then that implies that there exists at least one null value in that specific column. 

````python
state_insur.isna().sum()
````

In the table above, we can see that we have several columns that have `null` values specifically, 
- State Medicaid Expansion (2016)
- Medicaid Enrollment (2013)
- Medicaid Enrollment Change (2013-2016)

````python
# access the number of non-null values and data type of inputs per column
state_insur.info()
````

````python
# access counts of each unique value in State Medicaid Expansion (2016)
state_insur['State Medicaid Expansion (2016)'].value_counts()
````

Looking at the values of 'State Medicaid Expansion (2016)', we can see that the data is stored as a boolean value. We have also obtained the count of each boolean value in the dataset. From the table above, the mode of State Medicaid Expansion (2016) is `True`. We can then replace the single `null` value in `State Medicaid Expansion (2016)` with `True`.

````python
# replace all null values with True
state_insur['State Medicaid Expansion (2016)'].replace(np.nan, 'True', inplace = True)
````

This process is then repeated for the other columns with the `null` values. 

As we do not have real values for the `null` values, we chose to use the average values of the dataset as the replacement. This is a better alternative than deleting the entire entry as there is other data that can be extracted from the row. By using the average, we can also lower the chance of fluctuation of the average when we take the average of the data again.

````python
# obtain the mean of the column 'Medicaid Enrollment (2013)' and replace all null values with the mean
mean_of_medical_enrollment = state_insur['Medicaid Enrollment (2013)'].mean()
state_insur['Medicaid Enrollment (2013)'].replace(np.nan, mean_of_medical_enrollment, inplace = True)
````

````python
# obtain the mean of the column 'Medicaid Enrollment Change (2013-2016)' and replace all null values with the mean
mean_of_medical_enrollment_change = state_insur['Medicaid Enrollment Change (2013-2016)'].mean()
state_insur['Medicaid Enrollment Change (2013-2016)'].replace(np.nan, mean_of_medical_enrollment_change, inplace = True)
````

````python
## If data cleaning has been done properly,
# then there should be no null values left
state_insur.isna().sum()
````

As we can see, the sum of null values in each column is 0, implying that there no longer exists any null values. 

As the column State Medicaid Expansion (2016) stores boolean values as its data, we should change all values of `True` to be `1` and all values of `False` to be `0`. This is because it is much easier to graph integer values and it can come in handy when analyzing the data. 

````python
# changing True to 1 and False to 0
state_insur['State Medicaid Expansion (2016)'].replace('True', 1, inplace = True)
state_insur['State Medicaid Expansion (2016)'].replace('False', 0, inplace = True)
````

From the previous table above, we can see that there are columns that contain numerical values but are stored as strings. As a result, we need to convert these inputs into floats so that we may manipulate the values more easily when graphing the data. More specifically, we will convert all percentages into their decimal equivalent and store all other numerical values as floats. 

````python
# turning strings into floats and converting percentages into its decimal equivalent
state_insur['Uninsured Rate (2010)'] = (state_insur['Uninsured Rate (2010)'].str.rstrip('%').astype('float') / 100.0)
state_insur['Uninsured Rate (2015)'] = (state_insur['Uninsured Rate (2015)'].str.rstrip('%').astype('float') / 100.0)
state_insur['Uninsured Rate Change (2010-2015)'] = (state_insur['Uninsured Rate Change (2010-2015)'].str.rstrip('% ').astype('float') / 100.0)
state_insur['Average Monthly Tax Credit (2016)'] = (state_insur['Average Monthly Tax Credit (2016)'].str.lstrip('$').astype('float'))
state_insur.head()
````

At this point, we can conclude that data cleaning steps for the **State Health Insurance Coverage Data** dataset are now complete. We have done the following to this dataset
- Replace `null` values with the `mode` value for 3 columns
- Replace `boolean` values (`True`, `False`) with its integer representation for 1 column
- Changed cell values (represented by percentages) from `string` to `floats` in decimal value

By performing these steps, we are able to ensure that there are no missing values and structural errors are fixed. The rules stated above (defined under the `Data Cleaning` description) is fulfilled.

H0: Implementing socialized healthcare lowered or had no effect on the rate of increase in life expectancy in the United States. <br />

````python
# get the nataional average age at birth for each year
at_birth = n_mort[n_mort['Age']==0]
at_birth.tail()
````

I want only life expectancy at birth because I need to use just one age to use for my data analysis. I choose an age of 0 (at birth) because it is what is commonly referred to as life expectancy.
<br /><br />
Because life expectancy has gone up over time due to advances in medical care, it's impossible to determine the effect health insurances has on life expectancy. Because of this, we will see if the slope of a set of regression lines for the time before and after socialized healthcare was nationally implemented (2010) has different slopes. This might suggest that something that happened in 2010 made a significant difference to the life expectancy at birth. 

````python
p = sns.regplot(x='Year',y='ex',data=at_birth)
p.set_title('National Average Life Expectancy from 1959 to 2018')
p.set_ylabel('Life Expectancy (age)')
````

I am calculating and visualizing a regression over the entire timeframe to get an idea of the average increase in life expectancy at birth per year so that I have a baseline to compare our slopes from the subsets later. If we want to reject our null hypothesis we want our post-2010 calculated regression slope to be significantly greater than the slope of this line and the regression line in the following graph.

````python
#calculating the slope of the above using linear regression
slope0, _, r_value0, p_value0, std_err = linregress(at_birth['Year'], at_birth['ex'])
print('The average increase in life expectancy per year from 1959-2018 is', slope0)
````

Above I am calculating the slope of the line which gives the rate of change of life expectancy over time. The above number (~0.17) equates to an average increase in life expectancy of about 0.17 years of life per year in time. This rate represents the average increase in life expectancy caused by improvements in improvements in hygiene, policy, and advancements in medical care.

````python
print('the r-value is',r_value0)
print('the p-value is',p_value0)
````

Because the r-value is greater than 0.8, this graph represents a strong positive correlation between year and life expectancy. The p-value is much smaller than 0.01, so this slope represents a statistically significant deviation from a slope of zero.

Now we will be looking at the national average life expectancy within the span of 1990-2018, which is the time span we wish to address in our research. 

````python
sub_at_birth = at_birth[at_birth['Year']>=1990]
p1 = sns.regplot(x='Year',y='ex',data=sub_at_birth)
p1.set_title('National Average Life Expectancy from 1990 to 2018')
p1.set_ylabel('Life Expectancy (age)')
````

````python
#calculating the slope of the above using linear regression
slope1, _, r_value1, p_value1, std_err = linregress(sub_at_birth['Year'], sub_at_birth['ex'])
print('The average increase in life expectancy per year from 1990-2018 is', slope1)
````

Here we have calculated the rate of change for just the totality of our range. It is about 0.15 which represents an increase in life expectancy of about 0.15 years per year. This is slightly less than the average across our data set, but that could be caused by extraneous events related to cultural shifts in family values, or related to the declaration of war with Afghanistan in 2001. 

````python
print('the r-value is',r_value1)
print('the p-value is',p_value1)
````

This r-value represents a strong positive relationship, and the p-value says that this relationship is no longer statistically different from a slop of zero.

````python
p2 = sns.regplot(x='Year',y='ex',data=sub_at_birth[sub_at_birth['Year']<2010])
sns.regplot(x='Year',y='ex',data=sub_at_birth[sub_at_birth['Year']>=2010])
p2.set_title('National Average Life Expectancy from 1990 to 2018 (split at 2010)')
p2.set_ylabel('Life Expectancy (age)')
````

````python
sdf1 = sub_at_birth[sub_at_birth['Year']<2010]
sdf2 = sub_at_birth[sub_at_birth['Year']>=2010]

#calculating the slope of the above using linear regression before 2010
slope2a, _, r_value2, p_value2, std_err = linregress(sdf1['Year'], sdf1['ex'])

#calculating the slope of the above using linear regression after 2010
slope2b, _, r_value3, p_value3, std_err = linregress(sdf2['Year'], sdf2['ex'])

print('Before 2010, life expectancy rose by about', slope2a, 'per year.')
print('After 2010, life expectancy rose by about', slope2b, 'per year.')
````

The slopes are split in 2010 because that is when the Affordable Care Act was introduced. Because the rate of change decreases by a great deal after 2010, we cannot reject the null hypothesis. This does not mean that insurance doesn't have an effect. It just means that if it does have an effect, it cannot be seen by examining the rate of change in national mortality.

````python
print('Before 2010, the r-value is',r_value2)
print('Before 2010, the p-value is',p_value2,'\n')
print('After 2010, the r-value is',r_value3)
print('After 2010, the p-value is',p_value3)
````

The correlation is only strong before 2010. The p-values only show statistical significance before 2010. This further goes to show that we cannot reject the null hypothesis.

However, these results do not prove that there is a causal relationship between those variables; it only suggests that they are significantly related. There could be confounding factors like life expectancy increasing as medicine advances at the same time as socialized insurance infrastructure is made more accessible.
