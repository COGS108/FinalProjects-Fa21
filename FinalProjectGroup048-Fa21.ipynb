# Data Cleaning

### Cleaning Vaccination Trends Data:

This is what the original dataset for vaccination trends looks like: 

```python
# Dataset #1: vaccination trends data 

# Import vacc trends CSV 
df = pd.read_csv("vacc_trends_copy.csv")
df
```

#### Cleaning process:

Since we only plan on looking at the rate of COVID-19 vaccinations in California, we are only concerned about the columns with the date, the location, and the total doses administered that day. In the following cell, we removed all unneeded columns: 'Date type', '7-Day Avg Total Doses Daily', 'US 7-Day Total Doses Daily.'

We also noticed how every day has an additional row of information where the Date type is a 'report.' These rows have NaN for our column of interest and the rest have information that is irrelevant to our question so we will drop these rows later.

```python
# Remove unneeded columns: Date type, 7-Day Avg Total Doses Daily, US 7-Day Total Doses Daily 
df = df.drop(labels=['Date Type'], axis = 1)
df = df.iloc[:, :3]
df
```

Now, we can drop all the unnecessary rows that we noticed earlier. We do this by dropping all the rows with NaN.

```python
# Take out rows with missing values and reset the index
df = df.dropna().reset_index()
# df = df.iloc[:, 2:]

# Print out cleaned dataframe
df
```

We need to drop the last 2 rows to stay in our time period. Also, since our vaccination data is daily but our data for google searches is weekly, we need to convert our vaccination rates data into weekly data:

```python
# Convert vaccination rates data to weekly data
df['Date'] = df['Date'].astype('datetime64[ns]')
weekly_data = df.groupby("Location").resample('W-Sun', label='right', closed = 'right', on='Date').sum().reset_index().sort_values(by='Date')
weekly_data = weekly_data.drop(labels=['index'], axis = 1)

# Drop the last 2 rows to our designated time frame
weekly_data = weekly_data.iloc[:46, :]

weekly_data.head()
```

### Cleaning Interest in Travel Data:

Here is an example of an uncleaned dataset for the "daytrips" keyword. Notice there is an unnecessary header 'Category: All categories.' We will use skiprows when importing the CSVs to get rid of that.

```python
daytrips_original = pd.read_csv('cadaytrips.csv')
daytrips_original.head()
```

```python
# Uploading the CSV files for each search term 
# Use skiprows to skip the unnecessary header
daytrips = pd.read_csv('cadaytrips.csv', skiprows=2)
destinations = pd.read_csv('cadestinations.csv', skiprows=2)
roadtrips = pd.read_csv('caroadtrips.csv', skiprows=2)
sightseeing = pd.read_csv('casightseeing.csv', skiprows=2)
tourism = pd.read_csv('catourism.csv', skiprows=2)
travel = pd.read_csv('catravel.csv', skiprows=2)
trips = pd.read_csv('catrips.csv', skiprows=2)
vacations = pd.read_csv('cavacations.csv', skiprows=2)
local = pd.read_csv('localtravel.csv', skiprows=2)
places = pd.read_csv('placestogoinca.csv', skiprows=2)
traveltopic = pd.read_csv('traveltopic.csv', skiprows=2)
```

After we removed the unneccesary headers, we rename the columns for each search term data frame to reflect the correct search term:

```python
# Rename the columns of each search term data frame
daytrips = daytrips.rename({'california day trips: (California)':'california day trips'}, axis='columns')
destinations = destinations.rename({'california destinations: (California)':'california destinations'}, axis='columns')
roadtrips = roadtrips.rename({'california road trips: (California)':'california road trips'}, axis='columns')
sightseeing = sightseeing.rename({'california sightseeing: (California)':'california sightseeing'}, axis='columns')
tourism = tourism.rename({'california tourism: (California)':'california tourism'}, axis='columns')
travel = travel.rename({'california travel: (California)':'california travel'}, axis='columns')
trips = trips.rename({'california trips: (California)':'california trips'}, axis='columns')
vacations = vacations.rename({'california vacations: (California)':'california vacations'}, axis='columns')
local = local.rename({'local travel: (California)':'local travel'}, axis='columns')
places = places.rename({'places to go in california: (California)':'places to go in california'}, axis='columns')
traveltopic = traveltopic.rename({'Travel: (California)':'travel'}, axis='columns')
```

Now we combine the data from all the search terms into one dataframe so we can see all the data we want for when we analyze interest in travel. We also need to drop the last row so as to stay in our time period of analysis (12-20-2020 to 10-31-2021):

```python
all_travel = daytrips.join(destinations['california destinations']).join(roadtrips['california road trips']).join(sightseeing['california sightseeing']).join(tourism['california tourism']).join(travel['california travel']).join(trips['california trips']).join(vacations['california vacations']).join(local['local travel']).join(places['places to go in california']).join(traveltopic['travel'])

# Drop the last row so as to stay in our time frame
all_travel = all_travel.iloc[:46, :]
```

We add a sum column that adds the values in each row so we get a measurement for interest in general travel in California for the given week:

```python
# add a sum column
travel_sum = all_travel
sum_col = travel_sum['california day trips'] + travel_sum['california destinations'] + travel_sum['california road trips'] + travel_sum['california sightseeing'] + travel_sum['california tourism'] + travel_sum['california travel'] + travel_sum['california trips'] + travel_sum['california vacations'] + travel_sum['local travel'] + travel_sum['places to go in california'] + travel_sum['travel']
travel_sum['Travel Searches'] = sum_col
travel_sum.head()
```

Combining vaccination rate data and travel search data into one dataframe completes the cleaning process for this section.

```python
extracted_col = travel_sum["Travel Searches"]
vacc_trav = weekly_data
vacc_trav = vacc_trav.join(extracted_col)
vacc_trav = vacc_trav.rename(columns={'Total Doses Administered Daily':'total_doses','Travel Searches':'travel_searches'})
vacc_trav
```

### State parks

We follow the same process as above to clean our state parks dataset as it comes in the same format as the above travel search dataset. Below, you can see how we import the CSV's and drop the unnecessary header using skuprows, check for null vales, and then rename the columns with the appropriate key word.

```python
# Cleaning dataset state park datasets for the sub question

# Import respective state park CSV file
Emerald_bay = pd.read_csv("Emerald_bay.csv", skiprows=2)
Julia_pfeiffer = pd.read_csv("Julia_pfeiffer.csv", skiprows=2)
Point_lobos = pd.read_csv("Point_lobos.csv", skiprows=2)
Humboldt_redwoods = pd.read_csv("Humboldt_redwoods.csv", skiprows=2)
Burney_falls = pd.read_csv("Burney_falls.csv", skiprows=2)



# Checking for null vlaues
Emerald_bay = Emerald_bay.dropna()
Julia_pfeiffer = Julia_pfeiffer.dropna()
Point_lobos = Point_lobos.dropna()
Humboldt_redwoods = Humboldt_redwoods.dropna()
Burney_falls = Burney_falls.dropna()


# Rename the columns 
Emerald_bay = Emerald_bay.rename({'Emerald Bay State Park: (California)':'Emerald Bay State Park'}, axis='columns')
Julia_pfeiffer = Julia_pfeiffer.rename({'Julia Pfeiffer Burns State Park: (California)':'Julia Pfeiffer Burns State Park'}, axis='columns')
Point_lobos = Point_lobos.rename({'Point Lobos State Natural Reserve: (California)':'Point Lobos State Natural Reserve'}, axis='columns')
Humboldt_redwoods = Humboldt_redwoods.rename({'Humboldt Redwoods State Park: (California)':'Humboldt Redwoods State Park'}, axis='columns')
Burney_falls = Burney_falls.rename({'McArthur-Burney Falls Memorial State Park: (California)':'McArthur-Burney Falls Memorial State Park'}, axis='columns')


# There is no need to clean the data further since all rows and columns are essential
```

Now we combine the data from all the search terms into one dataframe so we can see all the data we want for when we analyze interest in travel to CA state parks. We also drop the last row so we stay in our time frame (ending October 31, 2021) and create a sum column that adds up the values for each week.

```python
state_parks = Emerald_bay.join(Julia_pfeiffer['Julia Pfeiffer Burns State Park']).join(Point_lobos['Point Lobos State Natural Reserve']).join(Humboldt_redwoods['Humboldt Redwoods State Park']).join(Burney_falls['McArthur-Burney Falls Memorial State Park'])

# Drop the last row so as to stay in our time frame
state_parks = state_parks.iloc[:46, :]

state_parks.head()
```

```python
s_park = state_parks
sum_col = s_park['Emerald Bay State Park'] + s_park['Julia Pfeiffer Burns State Park'] + s_park['Point Lobos State Natural Reserve'] + s_park['Humboldt Redwoods State Park'] + s_park['McArthur-Burney Falls Memorial State Park']
s_park['State Park Searches'] = sum_col
s_park.head()
```

```python
extracted_col = s_park["State Park Searches"]
vacc_stprk = weekly_data
vacc_stprk = vacc_stprk.join(extracted_col)
vacc_stprk = vacc_stprk.rename(columns={'Total Doses Administered Daily':'total_doses','State Park Searches':'stprk_searches'})
vacc_stprk
```

### Amusement Parks

We follow the same process to clean our amusement park data as we did above for the travel and state park datasets:

```python
# Cleaning Amusement park datasets for the sub question 

# Cleaning dataset for "disneyland" search term
disney = pd.read_csv("disneyland.csv")

# Checking for null vlaues
disney = disney.dropna() 

# rename columns
disney = disney.rename({'Category: All categories':'Week', 'Unnamed: 1':'Disneyland'}, axis='columns')

# dropping the first row 
disney = disney.iloc[1:, :]

# Follow the same cleaning process for the other amusement park search terms
six_flags = pd.read_csv("sixflags.csv")
six_flags = six_flags.dropna()
six_flags = six_flags.rename({'Category: All categories':'Week', 'Unnamed: 1':'Six Flags'}, axis='columns')
six_flags = six_flags.iloc[1:, :]

great_america = pd.read_csv("greatamerica.csv")
great_america = great_america.dropna()
great_america = great_america.rename({'Category: All categories':'Week', 'Unnamed: 1':'Great America'}, axis='columns')
great_america = great_america.iloc[1:, :]

legoland = pd.read_csv("legoland.csv")
legoland = legoland.dropna()
legoland = legoland.rename({'Category: All categories':'Week', 'Unnamed: 1':'Legoland'}, axis='columns')
legoland = legoland.iloc[1:, :]

universal = pd.read_csv("universalstudios.csv")
universal = universal.dropna()
universal = universal.rename({'Category: All categories':'Week', 'Unnamed: 1':'Universal Studios'}, axis='columns')
universal = universal.iloc[1:, :]
```

Now we combine the data from all the search terms into one dataframe so we can see all the data we want for when we analyze interest in travel to amusement parks.

At this point, we notice that something is wrong with our value types! We checked the datatypes using dtypes and saw that search popularity values were objects and not ints which we need them to be. So, we need to go ahead and convert the data types of all the search term columns from objects to ints:

```python
all_a_parks = disney.join(six_flags['Six Flags']).join(great_america['Great America']).join(legoland['Legoland']).join(universal['Universal Studios']).reset_index()
all_a_parks = all_a_parks.drop(columns=['index'])
all_a_parks.dtypes
```

```python
# Converting the datatypes of all the columns to ints so we can perform numerical operations
all_a_parks['Disneyland'] = all_a_parks['Disneyland'].astype(str).astype(int)
all_a_parks['Six Flags'] = all_a_parks['Six Flags'].astype(str).astype(int)
all_a_parks['Great America'] = all_a_parks['Great America'].astype(str).astype(int)
all_a_parks['Legoland'] = all_a_parks['Legoland'].astype(str).astype(int)
all_a_parks['Universal Studios'] = all_a_parks['Universal Studios'].astype(str).astype(int)


a_park = all_a_parks
sum_col = all_a_parks['Disneyland'] + all_a_parks['Six Flags'] + all_a_parks['Great America'] + all_a_parks['Legoland'] + all_a_parks['Universal Studios']
a_park['Amusement Park Searches'] = sum_col
a_park.dtypes
```

```python
extracted_col = all_a_parks["Amusement Park Searches"]
vacc_amprk = weekly_data
vacc_amprk = vacc_amprk.join(extracted_col)
vacc_amprk = vacc_amprk.rename(columns={'Total Doses Administered Daily':'total_doses','Amusement Park Searches':'amprk_searches'})
vacc_amprk.head()
```
