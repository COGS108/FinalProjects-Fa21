The best players from each conference are selected based on various metrics such as points per game, rebounds, and other statistical categories. Seeing as there has been an increase in arguments from those in the basketball community comparing various eras to one another, we thought that we would create a way to see whether or not as time has progressed, have the best players in the league gotten better.

# Hypothesis


We believe that as time has gone on the NBA All-Star team for the positions of PG, SG, and Center has become less difficult to make. There are various factors that led our group to come to this hypothesis. After viewing source #2, we learned that a championship 20 years ago was the hardest championship to win, meaning the league was overall more competitive. This may make a tighter and harder race for making the All-Star team as well for those past years rather than todays game. Also, in recent decades fan vote has been 50% of the criteria, giving fan favorites who might be less skilled a better chance at making the All-Star team.

# Setup

First, we need to import our libraries. For these data sets, we will need to use **Pandas**, **NumPy**, **Beautiful Soup**, the **urllib** module from Python, and **Seaborn**.

```python
## Import libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from urllib.request import urlopen
import seaborn as sns
import matplotlib.pyplot as plt
```

## Player Stats
When pulling the data, we need to account for each year. So, we loop through the years since the All-Star game has existed (1951-Present).

For each year, we go to the URL of our data set and scrape the table data using Beautiful Soup. All the data online is stored using the table element in HTML. Using Beautiful Soup, we can select the observations by finding the table rows (*tr*), then finding the table data for each row (*td*).

For simplicity, we've created a dictionary where each key in it is a year and the value for that key is the data frame of player stats for that year.

**NOTE: This is a large data set and will take a minute or two to load!** 

```python
## Pulling player stats

print("Loading", end="")

df = dict() # Dictionary to hold data frame for each year
for year in range(1951, 2021):
    print(".", end="")
    
    # Opening url to feed Beautiful Soup
    url = "https://www.basketball-reference.com/leagues/NBA_{}_totals.html".format(year)
    html = urlopen(url)
    soup = BeautifulSoup(html)
    
    cols = list() # To hold our column categories
    
    # Here, we findAll by tr and limit our search to 1 row. This will 
    # give us the top row containing the column categories.
    # We then findAll by th to retrieve the actual categories and store them into cols
    for element in soup.findAll('tr', limit=1)[0].findAll('th'):
        if element.getText() != "Rk":
            cols.append(element.getText())
    
    rows = soup.findAll('tr')[1:] # Finds all rows except for first (categories)
    
    stats = list() # Stats list to hold player data
    
    # Here, we iterate through each row creating a new list that represents player stats
    for i in range(len(rows)):
        stats.append(list())
        
        # We then retrieve the data for each player and store them into the list within stats
        for element in rows[i].findAll('td'):
            stats[i].append(element.getText())
            
    df[year] = pd.DataFrame(stats, columns=cols) # We assign a key of the year with the data frame of stats for that year
    
print("Done!")
```

## All-Star Team Rosters

Next, we needed all the names for the All-Star team rosters for each season.

To achieve this, we used the same tools and format as before. The only difference is that there is a missing dataset for the year of 1999, so we will skip that one.

```python
## Pulling All-Star rosters

print("Loading", end="")

allstars = dict() # Dictionary to hold rosters for each year
for year in range(1951, 2021):
    print(".", end="")
    
    # Skip year of 1999
    if year == 1999:
        continue
    
    # Opening url to feed Beautiful Soup
    url = "https://www.basketball-reference.com/allstar/NBA_{}.html".format(year)
    html = urlopen(url)
    soup = BeautifulSoup(html)
    
    names = list() # List to store the names of players
    
    # Here, there are two tables (2 sides of All-Star game), and we need to check both.
    # Table bodies can be found using tbody.
    for element in soup.findAll('tbody', limit=2):
        
        # We then iterate through each row (tr)
        for row in element.findAll('tr'):
            
            # Find the name of each player (th)
            # This will also pull in categorical column names unnecessarily which we will need to remove later
            for item in row.findAll('th'):
                names.append(item.getText())
                
    allstars[year] = pd.DataFrame(names, columns={'player'}) # We assign a key of the year with the list of All-Star players for that year

print("Done!")
```

# Data Cleaning

## Player stats dataset
### 1. Missing data
Starting with the main set of data, the player stats, we saw that for earlier years there were important stats missing such as 3 point goals, blocks, and steals. We found these types of stats to be critical for the positions we chose to analyze. Therefore, we decided to drop any observations that did not have data for these stats to better our comparisons between players for each season.

However, removing this data isn't that simple. If we simply *dropna()* we will remove more data than necessary because there are other categories that weren't collected in earlier years. Some of these categories we have deamed as irrelevant; for example, the category ***GS*** or games started. Additionally, even when there is data for 3 point goals, if the amount made is 0 and the amount attempted is 0, it gives us a null value for that observation's 3 point percentage.

To remedy this issue, we will remove any irrelevant columns from the data frames, fill in any null percentage stats with a percentage of .000, and finally, drop observations containing null data. This will ensure that we don't unnecessarily remove any observations that we actually have data for.

### 2. Symbols in names
We noticed some of the names in the tables contained asterisks at the end. This isn't good because when pulling such players fromt the lists of All-Star teams, we noticed there were no asterisks. In order to keep the data consistent, we will be removing asterisks from such players' names.

```python
## Function to remove asterisks from names
def standardize_name(str_in):    
    str_in = str(str_in) # Cast as string just in case
    str_in = str_in.strip() # Strip whitespace

    # Replace * with null character
    if '*' in str_in:
        str_in = str_in.replace('*', '')
        output = str_in;
    else:
        output = str_in
    return output
```

```python
assert(standardize_name("Lebron James*") == "Lebron James")
assert(standardize_name("James Harden") == "James Harden")
```

### 3. Duplicates
Another issue with this dataset is that there are a ton of duplicates. This is because during the year a player may have switched teams and so they have 2-3 sets of stats. However, the dataset contains the player's aggregated data for the season before the separated data they had for each team. This means all we have to do is remove duplicates and select to keep the first appearance of the data in order to keep only the total stats from the entire season.

### 4. Data Type
The final issue with this dataset is that everything was imported as a string. We need to make sure each category has the correct data type. To do this, we will loop through the columns and change their data type accordingly.

```python
## Cleaning player stats data
for year in range(1951, 2021):
    
    # Drop irrelevant categories
    df[year] = df[year].drop(columns='GS')
    
    df[year]['Player'] = df[year]['Player'].apply(standardize_name) # Removing asterisks and whitespace from names
    
    # Simultaneously changing data types and filling empty data
    for col in df[year].columns:
        current = df[year][col]
        if current.name != 'Player' and current.name != 'Pos' and current.name != 'Tm':
            if '%' in current.name:
                df[year][col] = pd.to_numeric(current, errors='coerce', downcast='float').fillna(.000)
            else:
                df[year][col] = pd.to_numeric(current, errors='coerce', downcast='integer')
                
    df[year] = df[year].dropna() # Dropping any rows with actual null data
    
    # Need to go back and re-cast integers as they get cast as floats for null values
    for col in df[year].columns:
        current = df[year][col]
        if current.name != 'Player' and current.name != 'Pos' and current.name != 'Tm':
            if '%' not in current.name:
                df[year][col] = pd.to_numeric(current, errors='coerce', downcast='integer')
                
    df[year] = df[year].dropna() # Dropping any rows with actual null data
    
    df[year] = df[year].drop_duplicates(subset='Player', keep='first') # Removing duplicate players
    df[year] = df[year].reset_index(drop=True) # Index formatting
    
```

```python
# Asserts for sanity check
for year in range(1951, 2021):
    assert('GS' not in df[year].columns) # Check for removed GS column
    
    assert(True not in df[year].duplicated(subset='Player').to_list()) # Check for duplicate players
    
    for col in df[year].columns:
        current = df[year][col]
        assert(True not in current.isna().to_list()) # No null values
        if current.name != 'Player' and current.name != 'Pos' and current.name != 'Tm':
            if '%' in current.name:
                assert(np.issubdtype(current.dtype, np.floating)) # % columns are floats
            else:
                assert(np.issubdtype(current.dtype, np.integer)) # All other columns are ints
```

## Analysis
### Overview

### 2. Setting up calculation
In order to calculate the relative skill of players we adapted a formula that utilizes season averages and weights on different stats. For each position, PG, SG, and C, we came up with weights for each stat where stats that were more relevant to the given position were weighted higher.

Before we get into our formula, we still need to calculate each season's stat averages.

```python
# Calulating season averages

# New data frame for years averages
avgs = pd.DataFrame()
avgs['Year'] = [x for x in range(1980,2021)]

omit = ['Name', 'Pos', 'Year', 'All-Star'] # Columns to omit in new data frame

# Cycle through each column for each year, find mean for that season, & store in data frame
for year in range(1980, 2021):
    current = stats_ref[stats_ref['Year'] == year]
    for col in current.columns:
        if col not in omit:
            mean = current[col].mean()
            avgs.loc[avgs['Year'] == year, col] = mean
```

Now that we have setup the table of season averages, we can now move on to setting up our formula.

For this formula, we take each stat from each player from each season and divide it by the season average, then multiply it by our designated weights. For example, if for a point guard in a given season his points are 500 and the season average is 400 points, and we weigh total points at 10% for point guards the calculation for that stat will be 10*(500/400). We will perform this for all the stats and total them to get our overall score for that player.

Looking at only PG, SG, and C positions, we need to define the weights for each position. To determine these weights we used our own best judgement with the knowledge that we have and researched what skills recruiters look for in all of these positions.

Our weights are as follows:
- **Point Guard**
    - *Most Important*: Points, Turnovers, Assists, 3P, 3P%
        - 45% TOTAL WEIGHT or 9% PER STAT
    - *Middle Important*: Steals, FT, FT%, FG, FG%, eFG%, 2P, 2P%
        - 48% TOTAL WEIGHT or 6% PER STAT 
    - *Least Important*: Blocks, Rebounds, MP
        - 7% TOTAL WEIGHT or 2.33% PER STAT
        
- **Shooting Guard**
    - *Most Important*: Points, FG, FG%, eFG%
        - 32% TOTAL WEIGHT or 8% PER STAT
    - *Middle Important*: FT%, FT, 3P%, 3P, 2P, 2P%, Rebounds
        - 42% TOTAL WEIGHT or 6% PER STAT
    - *Least Important*: Steals, Assists, Blocks, MP, Turnovers
        - 26% TOTAL WEIGHT or 5.2% PER STAT

- **Center**
    - *Most Important*: Blocks, Rebounds, 2P, 2P%
        - 40% TOTAL WEIGHT or 10% PER STAT
    - *Middle Important*: Turnovers, FT, FT%, FG, FG%, Points
        - 36% TOTAL WEIGHT or 6% PER STAT
    - *Least Important*: Steals, 3P, 3P%, Assists, eFG%, MP
        - 24% TOTAL WEIGHT or 4% PER STAT

### 3. Performing Calculation
Now that we have determined the wieghts we will calculate the scores of each PG, SG, & C, then plot each position's scores with respect to time to determine a conclusion.

```python
# Calculation for PG

pgs = stats_ref[stats_ref['Pos'].str.contains('PG')] # List of PGs

# Creating weights for each stat
pg_weights = dict()
for col in pgs.columns:
    if col != 'Name' and col != 'Year' and col != 'All-Star' and col != 'Pos':
        if col == 'Blocks' or col == 'Rebounds' or col == 'MP':
            pg_weights[col] = 2.33
        elif col == 'Points' or col == 'Turnovers' or col == 'Assists' or col == '3P' or col == '3P%':
            pg_weights[col] = 9.0
        else:
            pg_weights[col] = 6.0

# Calculating the overall scores and storing to list
totals = list()
for index, row in pgs.iterrows():
    total = 0
    year = row['Year']
    avg = avgs[avgs['Year'] == year].reset_index()
    for col in avgs.columns:
        if col != 'Year':
            # Need to subtract for turnovers since it is a bad stat to have high
            if col == 'Turnovers':
                total -= pg_weights[col]*(row[col]/avg.iloc[0][col])
            else:
                total += pg_weights[col]*(row[col]/avg.iloc[0][col])
    totals.append(round(total, 2))
    
pgs['Total'] = totals # Assign scores to players
pgs
```

```python
# Calculation for SG

sgs = stats_ref[stats_ref['Pos'].str.contains('SG')] # List of SGs

# Creating weights for each stat
sg_weights = dict()
for col in sgs.columns:
    if col != 'Name' and col != 'Year' and col != 'All-Star' and col != 'Pos':
        if col == 'Blocks' or col == 'Steals' or col == 'MP' or col == 'Assists' or col == 'Turnovers':
            sg_weights[col] = 5.2
        elif col == 'Points' or col == 'FG' or col == 'FG%'or col == 'eFG%':
            sg_weights[col] = 8.0
        else:
            sg_weights[col] = 6.0
            
# Calculating the overall scores and storing to list
totals = list()
for index, row in sgs.iterrows():
    total = 0
    year = row['Year']
    avg = avgs[avgs['Year'] == year].reset_index()
    for col in avgs.columns:
        if col != 'Year':
            # Need to subtract for turnovers since it is a bad stat to have high
            if col == 'Turnovers':
                total -= sg_weights[col]*(row[col]/avg.iloc[0][col])
            else:
                total += sg_weights[col]*(row[col]/avg.iloc[0][col])
    totals.append(round(total, 2))

sgs['Total'] = totals # Assign scores to players
sgs
```

```python
# Calculation for C

cs = stats_ref[stats_ref['Pos'].str.contains('C')] # List of Cs

# Creating weights for each stat
c_weights = dict()
for col in cs.columns:
    if col != 'Name' and col != 'Year' and col != 'All-Star' and col != 'Pos':
        if col == 'Blocks' or col == 'Rebounds' or col == '2P' or col == '2P%':
            c_weights[col] = 10.0
        elif col == 'Steals' or col == 'MP' or col == 'Assists' or col == '3P' or col == '3P%' or col == 'eFG%':
            c_weights[col] = 4.0
        else:
            c_weights[col] = 6.0

# Calculating the overall scores and storing to list
totals = list()
for index, row in cs.iterrows():
    total = 0
    year = row['Year']
    avg = avgs[avgs['Year'] == year].reset_index()
    for col in avgs.columns:
        if col != 'Year':
            # Need to subtract for turnovers since it is a bad stat to have high
            if col == 'Turnovers':
                total -= c_weights[col]*(row[col]/avg.iloc[0][col])
            else:
                total += c_weights[col]*(row[col]/avg.iloc[0][col])
    totals.append(round(total, 2))

cs['Total'] = totals # Assign scores to players
cs
```
